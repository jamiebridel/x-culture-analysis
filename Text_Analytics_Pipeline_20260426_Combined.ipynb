```python
import json
import matplotlib.pyplot as plt      # Added for plotting
import numpy as np
import os
import pandas as pd                  # Used for DataFrame operations
import re                            # For regex to extract year from filenames
import spacy                         # Added for NationalityExtractor
import sys
import warnings
from abc import ABC, abstractmethod
from contextlib import contextmanager
from datetime import datetime
from google.colab import drive
from IPython.display import display                  # Explicitly import display
from matplotlib.backends.backend_pdf import PdfPages # Added for multi-page PDF output
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_distances

# Instantiate Config class for global variables
class Config:
    def __init__(self):
        self.input_data_dir = "/content/drive/MyDrive/x-culture/01_Data/"
        self.OUTPUT_FOLDER = '/content/drive/MyDrive/x-culture/02_Results_Data/'
        self.CODEBOOK_FILE = f"{self.input_data_dir}Comments_CodeBook_v4.xlsx"
        self.COMMENTS_FILE = f"{self.input_data_dir}Comments_Data_All.xlsx"
        self.LMS_LEXICON_FILE = f"{self.input_data_dir}lexicons_LSM.json"
        self.LIWC_LEXICON_FILE = f"{self.input_data_dir}lexicons_LIWC.json"
        self.SUCC_FAIL_LEXICON_FILE = f"{self.input_data_dir}lexicons_success_fail.json"

        # --- Define the semester year dynamically (or 'All') ---
        _raw_semester_year_setting = ['2019']
        # _raw_semester_year_setting = 'All'

        if isinstance(_raw_semester_year_setting, str) and _raw_semester_year_setting != 'All':
            self.semester_year = [_raw_semester_year_setting] # Convert single year string to list
        elif isinstance(_raw_semester_year_setting, list):
            self.semester_year = [str(y) for y in _raw_semester_year_setting] # Ensure list elements are strings
        else: # Must be 'All'
            self.semester_year = _raw_semester_year_setting

        # --- Configuration variable for redirecting warnings ---
        self.REDIRECT_WARNINGS_TO_FILE = False
        self.warnings_log_file = os.path.join(self.OUTPUT_FOLDER, 'python_warnings_log.txt')

        # --- New control flags for analysis execution ---
        self.run_individual_analysis = True
        self.run_team_analysis = False

class OSDriveCheck:
    def check_and_mount_drive(self):
        if not os.path.exists('/content/drive'):
            print('Google Drive is not mounted. Mounting ...')
            drive.mount('/content/drive/')
        else:
            print('Google Drive is already mounted.')


@contextmanager
def redirect_stderr_to_file(filepath):
    old_stderr = sys.stderr
    with open(filepath, 'w') as f:
        sys.stderr = f
        try:
            yield
        finally:
            sys.stderr = old_stderr


# Abstract base class for Calculating Text Metrics
class TextMetric(ABC):
    @abstractmethod
    def compute_metric(self, team_week_df):
        pass

# Abstract base class for Data Loading
class DataLoader(ABC):
    @abstractmethod
    def load_data(self, file_path):
        pass

# Abstract base class for Exporting to specified formats
class Exporter(ABC):
    @abstractmethod
    def export(self, df_to_export, filename_suffix="", **kwargs):
        pass

# New abstract base class for Plots
class Plotter(ABC):
    @abstractmethod
    def plot(self, data_df, output_folder):
        pass

# Abstract base class for validation checks
class ValidationCheck(ABC):
    @abstractmethod
    def validate(self):
        pass

# New utility class for team-related operations
class TeamUtils:
    @staticmethod
    def parse_team_id_components(team_id_series):
        # Convert 'team_id' to string type to prevent AttributeError if it contains non-string values
        team_id_str = team_id_series.astype(str)
        cohort = team_id_str.apply(lambda x: '-'.join(x.split('-')[1:3]) if x != 'nan' else np.nan)
        year = team_id_str.apply(lambda x: x.split('-')[0] if x != 'nan' else np.nan)
        return cohort, year


class TeamLSMWeekValidation(ValidationCheck):
    def __init__(self, wide_lsm_metrics_df):
        self.df = wide_lsm_metrics_df

    def validate(self):
        print("\n--- Check: Valid Team LSM Score Presence (Weeks 1 & 2) ---")

        # Ensure the columns for team LSM composite score for week 1 and week 2 exist
        col_week1 = 'lsm_week_1'
        col_week2 = 'lsm_week_2'

        has_week1_score = self.df[col_week1].notna() if col_week1 in self.df.columns else pd.Series(False, index=self.df.index)
        has_week2_score = self.df[col_week2].notna() if col_week2 in self.df.columns else pd.Series(False, index=self.df.index)

        # Categorize teams
        valid_week1_only = self.df[has_week1_score & ~has_week2_score].shape[0]
        valid_week2_only = self.df[~has_week1_score & has_week2_score].shape[0]
        valid_neither_week = self.df[~has_week1_score & ~has_week2_score].shape[0]
        valid_both_weeks = self.df[has_week1_score & has_week2_score].shape[0]

        # Create a table for visualization
        validation_results = pd.DataFrame({
            'Category': ['Valid in Week 1 Only', 'Valid in Week 2 Only', 'Valid Neither Week 1 Nor 2', 'Valid in Both Weeks'],
            'Number of Teams': [valid_week1_only, valid_week2_only, valid_neither_week, valid_both_weeks]
        })

        print(validation_results.to_string())
        print("------------------------------------------------------------------\n")

# New class for LSM Analyzer
class LSMAnalyzer(ValidationCheck):
    def __init__(self, wide_lsm_metrics_df):
        self.df = wide_lsm_metrics_df

    def validate(self):
        self.analyze_lsm_trends()

    def analyze_lsm_trends(self):
        print("\n--- Analyzing Team LSM Trends ---")

        # 1. Identify all columns that start with 'lsm_week_'
        lsm_week_cols = [col for col in self.df.columns if col.startswith('lsm_week_')]

        if not lsm_week_cols:
            print("No 'lsm_week_' columns found for analysis.")
            return

        # 2. Calculate and print overall descriptive statistics
        print("\nOverall LSM Statistics Across All Weeks:")
        overall_stats = self.df[lsm_week_cols].describe().loc[['mean', 'std', 'min', 'max']]
        print(overall_stats.to_string())

        # 3. Group by 'cohort' and calculate descriptive statistics
        if 'cohort' in self.df.columns:
            print("\nLSM Statistics Grouped by Cohort:")
            cohort_stats = self.df.groupby('cohort')[lsm_week_cols].describe().loc[:, (slice(None), ['mean', 'std', 'min', 'max'])]
            print(cohort_stats.to_string())
        else:
            print("Cohort column not found. Skipping cohort-wise LSM analysis.")

        print("-------------------------------------\n")


class CorrelationAnalyzer(ValidationCheck):
    def __init__(self, wide_metrics_df):
        self.df = wide_metrics_df
        self.correlation_results_list = [] # List to store correlations for each week
        self.all_correlations_df = pd.DataFrame() # DataFrame to store aggregated results

    def validate(self):
        self.analyze_lsm_correlations()

    def analyze_lsm_correlations(self):
        print("\n--- Analyzing Team LSM Correlations with Other Metrics ---")

        # Ensure 'cohort' and 'year' columns exist in the DataFrame
        if 'cohort' not in self.df.columns or 'year' not in self.df.columns:
            print("Warning: 'cohort' or 'year' column not found in wide_metrics_df. Cannot perform year/cohort-wise analysis.")
            grouped_dfs = [('Overall', self.df)] # Treat as one overall group
        else:
            # Group by year and cohort
            grouped_dfs = self.df.groupby(['year', 'cohort'])

        for group_key, group_df in grouped_dfs:
            year_name = group_key[0] if isinstance(group_key, tuple) else 'Overall'
            cohort_name = group_key[1] if isinstance(group_key, tuple) else group_key
            print(f"\n--- Correlations for Year: {year_name}, Cohort: {cohort_name} ---")

            lsm_cols = [col for col in group_df.columns if col.startswith('lsm_week_')]
            other_metric_prefixes = [
                'coordination_week_', 'participation_equality_week_', 'novelty_week_',
                'identity_we_i_week_', 'safety_week_', 'task_conflict_week_', 'relationship_conflict_week_',
                'insight_week_', 'causation_week_'
            ]

            all_week_cols = [col for col in group_df.columns if '_week_' in col]
            weeks = sorted(list(set([int(col.split('_')[-1]) for col in all_week_cols if col.split('_')[-1].isdigit()])))

            if not weeks:
                print(f"No weekly metric columns found for correlation analysis in Year {year_name}, Cohort {cohort_name}.")
                continue

            for week in weeks:
                current_week_lsm_col = f'lsm_week_{week}'

                if current_week_lsm_col not in group_df.columns or group_df[current_week_lsm_col].isnull().all():
                    continue

                current_week_other_metrics = []
                for prefix in other_metric_prefixes:
                    col_name = f'{prefix}{week}'
                    if col_name in group_df.columns:
                        current_week_other_metrics.append(col_name)

                if not current_week_other_metrics:
                    continue

                cols_for_corr = [current_week_lsm_col] + current_week_other_metrics
                week_df_subset = group_df[cols_for_corr].dropna()

                if week_df_subset.empty or len(week_df_subset) < 2:
                    continue

                correlation_matrix = week_df_subset.corr(method='pearson')

                lsm_correlations_with_others = correlation_matrix.loc[[current_week_lsm_col], current_week_other_metrics]

                # Process and store the correlations
                corr_df_week = lsm_correlations_with_others.T.reset_index()
                corr_df_week.columns = ['metric_name_full', 'correlation_with_lsm']
                corr_df_week['year'] = year_name # Add year information
                corr_df_week['cohort'] = cohort_name # Add cohort information
                corr_df_week['week'] = week
                corr_df_week['metric_name'] = corr_df_week['metric_name_full'].apply(lambda x: '_'.join(x.split('_')[:-2]))
                corr_df_week = corr_df_week[['year', 'cohort', 'week', 'metric_name', 'correlation_with_lsm', 'metric_name_full']]
                self.correlation_results_list.append(corr_df_week)

                # Display to console (original behavior)
                print(f"\nCorrelations for Week {week}:")
                display(lsm_correlations_with_others)
                plt.close('all')

        if self.correlation_results_list:
            self.all_correlations_df = pd.concat(self.correlation_results_list, ignore_index=True)
            print("\nAggregated Year and Cohort-wise Correlation Results Preview:")
            display(self.all_correlations_df.head())
        else:
            self.all_correlations_df = pd.DataFrame(columns=['year', 'cohort', 'week', 'metric_name', 'correlation_with_lsm', 'metric_name_full'])
            print("\nNo year and cohort-wise correlations found to aggregate.")

        print("----------------------------------------------------------\n")

    def get_all_correlations_df(self):
        return self.all_correlations_df


# DataLoader implementation for Excel data loading
class ExcelDataLoader(DataLoader):
    def load_data(self, file_path):
        return pd.read_excel(file_path, engine='openpyxl')


class RawCommentsDataLoader:
    def __init__(self, input_data_dir, semester_year, data_loader: DataLoader):
        self.input_data_dir = input_data_dir
        self.semester_year = semester_year
        self.data_loader = data_loader

    def load_raw_comments_data(self):
        all_data_file_path = os.path.join(self.input_data_dir, "Comments_Data_All.xlsx")
        all_comments_data_df = pd.DataFrame() # Initialize empty DataFrame for all comments

        try:
            print(f"Attempting to load 'Comments_Data_All.xlsx' from {all_data_file_path}...")
            all_comments_data_df = self.data_loader.load_data(all_data_file_path)
            print(f"Successfully loaded base data from 'Comments_Data_All.xlsx'. Total rows: {len(all_comments_data_df)}")
        except FileNotFoundError:
            print(f"Error: 'Comments_Data_All.xlsx' not found at {all_data_file_path}. Returning empty DataFrame.")
            return pd.DataFrame()
        except Exception as e:
            print(f"An error occurred while loading 'Comments_Data_All.xlsx': {e}. Returning empty DataFrame.")
            return pd.DataFrame()

        # If loading failed, return empty DataFrame
        if all_comments_data_df.empty:
            return pd.DataFrame()

        if self.semester_year == 'All':
            print("Processing for all years.")
            return all_comments_data_df
        elif isinstance(self.semester_year, list):
            years_to_filter = [int(y) for y in self.semester_year]
            print(f"Filtering data for years {self.semester_year}...")
            # Use the new utility method for parsing year
            _, all_comments_data_df['year'] = TeamUtils.parse_team_id_components(all_comments_data_df['team_id'])
            all_comments_data_df['year'] = pd.to_numeric(all_comments_data_df['year'], errors='coerce')
            filtered_data = all_comments_data_df[all_comments_data_df['year'].isin(years_to_filter)].copy()
            filtered_data = filtered_data.drop(columns=['year']) # Drop temporary 'year' column
            print(f"Successfully filtered data for years {self.semester_year}. Total rows: {len(filtered_data)}")
            return filtered_data
        else:
            print(f"Invalid semester_year type from Config: {type(self.semester_year)}. Expected 'All' or a list of years. Returning empty DataFrame.")
            return pd.DataFrame()

# Definition for CodebookProcessor (formerly CodebookMapper) - UPDATED
class CodebookProcessor:
    def __init__(self, codebook_df, raw_comments_data_df):
        self.codebook_df = codebook_df
        self.raw_comments_data_df = raw_comments_data_df

    def process_codebook_and_data(self):
        data = self.raw_comments_data_df.copy()
        codebook = self.codebook_df.copy()

        # --- 1) Filter codebook to only include text-based questions ---
        codebook_vars = codebook[codebook["Variable Name"].notna()].copy()

        # Only keep codebook entries that match actual dataset columns
        codebook_vars = codebook_vars[codebook_vars["Variable Name"].isin(data.columns)]

        # --- FIX: Handle non-numeric 'Week' values by extracting the number ---
        # Apply a regex to extract digits, then convert to numeric
        codebook_vars["Week"] = pd.to_numeric(
            codebook_vars["Week"].astype(str).str.extract(r'(\d+)').squeeze(),
            errors="coerce"
        )

        # Keep only rows with week info (drop NA weeks) - now applied after extraction
        codebook_vars = codebook_vars.dropna(subset=["Week"])

        # --- 2) Build list of ALL text fields present in the data based on codebook mapping ---
        text_fields = codebook_vars["Variable Name"].tolist()

        # --- 3) Melt/reshape the dataset using *ONLY* the mapped text fields ---
        long_df = (
            data.melt(
                id_vars=["student_id", "team_id", "semester_id"],
                value_vars=text_fields,
                var_name="variable_name",
                value_name="text"
            )
            .dropna(subset=["text"])
        )

        # --- 4) Merge week/question metadata from the codebook ---
        long_df = long_df.merge(
            codebook_vars[["Variable Name", "Original Question", "Week"]],
            left_on="variable_name",
            right_on="Variable Name",
            how="left"
        )

        # Cleanup
        long_df = long_df.drop(columns=["Variable Name"])
        long_df = long_df.rename(columns={'Original Question': 'question', 'Week': 'week'})

        # --- 5) Basic cleaning ---
        long_df["text"] = (
            long_df["text"]
            .astype(str)
            .str.replace("\n", " ", regex=False)
            .str.strip()
        )
        print(f"Final long_df has {len(long_df)} entries after processing.")
        return long_df


# New Class to load Dictionaries.
class LexiconLoader:
    # Modified constructor to accept two lexicon file paths
    def __init__(self, lsm_lexicon_file_path, liwc_fw_file_path, succ_fail_fw_file):
        # Load general lexicon data (hedges, apologies, etc.)
        with open(lsm_lexicon_file_path, 'r') as f:
            lsm_lexicon_data = json.load(f)

        # Load Success-Fail function word data
        with open(succ_fail_fw_file, 'r') as f:
            succ_fail_lexicon_data = json.load(f)

        # Load LIWC function word data
        with open(liwc_fw_file_path, 'r') as f:
            liwc_lexicon_data = json.load(f)

        self.fw_dict = {}
        if 'function_words' in liwc_lexicon_data and isinstance(liwc_lexicon_data['function_words'], dict):
            for category_name, category_content in liwc_lexicon_data['function_words'].items():
                if isinstance(category_content, list):
                    self.fw_dict[category_name] = set(word.lower() for word in category_content)
                elif isinstance(category_content, dict):
                    for sub_category_name, sub_category_words in category_content.items():
                        if isinstance(sub_category_words, list):
                            self.fw_dict[sub_category_name] = set(word.lower() for word in sub_category_words)

        self.hedges = lsm_lexicon_data.get('hedges', [])
        self.apologies = lsm_lexicon_data.get('apologies', [])
        self.task_words = lsm_lexicon_data.get('task_words', [])
        self.rel_words = lsm_lexicon_data.get('rel_words', [])
        self.insight_words = lsm_lexicon_data.get('insight_words', [])
        self.causation_words = lsm_lexicon_data.get('causation_words', [])
        self.success_set = succ_fail_lexicon_data.get('success_phrases', [])
        self.failure_set = succ_fail_lexicon_data.get('failure_phrases', [])


    def display_fw_summary(self):
        for category, words_list in self.fw_dict.items():
            print(f"  {category.capitalize()}: {', '.join(list(words_list)[:10])}{'...' if len(words_list) > 10 else ''}")
        for success_words in self.success_set:
            print(success_words)
        for failure_words in self.failure_set:
            print(f"  Failure: {(failure_words)}")


class LSMCalculator(TextMetric):
    def __init__(self, fw_dict):
        self.fw_dict = fw_dict

    def _count_fw(self, text, lexicon):
        text_str = str(text)
        if text_str.lower() == 'nan' or not text_str.strip():
            return np.nan
        words = text_str.lower().split()
        if not words:
            return np.nan
        return sum(w in lexicon for w in words) / (len(words) + 1)

    def compute_metric(self, team_week_df):
        if team_week_df['student_id'].nunique() < 3 or len(team_week_df) <= 1:
            return np.nan

        results = {}
        df_tw_copy = team_week_df.copy()
        for cat, lexicon in self.fw_dict.items():
            df_tw_copy[cat] = df_tw_copy['text'].apply(lambda x: self._count_fw(x, lexicon))
            avg = df_tw_copy[cat].mean()
            diff = abs(df_tw_copy[cat] - avg).mean()
            results[cat] = 1 - (diff / (avg + 1e-6))
        return np.mean(list(results.values()))

class CoordinationCalculator(TextMetric):
    def compute_metric(self, team_week_df):
        df_tw_copy = team_week_df.copy()

        df_tw_copy['length'] = df_tw_copy['text'].apply(lambda x: len(str(x).split()) if str(x).strip().lower() != 'nan' else 0)

        if df_tw_copy['length'].sum() == 0:
            return np.nan

        dominant = df_tw_copy.loc[df_tw_copy['length'].idxmax()]

        dominant_text_str = str(dominant['text'])
        if dominant_text_str.lower() == 'nan' or not dominant_text_str.strip():
            return np.nan

        dom_words = set(dominant_text_str.lower().split())
        def score(text):
            text_str = str(text)
            if text_str.lower() == 'nan' or not text_str.strip():
                return np.nan
            w = set(text_str.lower().split())
            if not w:
                return np.nan
            return len(w & dom_words) / (len(w) + 1)
        return np.nanmean(df_tw_copy['text'].apply(score))

class IdentityCalculator(TextMetric):
    def compute_metric(self, team_week_df):
        def compute_identity_single(text):
            text_str = str(text)
            if text_str.lower() == 'nan' or not text_str.strip():
                return np.nan
            t = ' ' + text_str.lower() + ' '
            I = t.count(' i ')
            W = t.count(' we ')

            if len(text_str.split()) == 0:
                return np.nan

            return W / (I + 1)
        # Use np.nanmean to handle potential NaN scores from individual text processing
        return np.nanmean(team_week_df['text'].apply(compute_identity_single))

class SafetyCalculator(TextMetric):
    def __init__(self, hedges, apologies):
        self.hedges = hedges
        self.apologies = apologies

    def compute_metric(self, team_week_df):
        def compute_safety_single(text):
            text_str = str(text)
            if text_str.lower() == 'nan' or not text_str.strip():
                return np.nan
            t = text_str.lower()
            q = t.count('?')
            h = sum(t.count(hg) for hg in self.hedges)
            a = sum(t.count(ap) for ap in self.apologies)
            return q + h + a
        return np.nanmean(team_week_df['text'].apply(compute_safety_single))

# --------
# New Class to calculate perceived effectiveness.
# --------
class PerceivedEffectiveness(TextMetric):
    def __init__(self, success_set, failure_set):
        self.success_set = success_set
        self.failure_set = failure_set

        keyword_matcher: Optional[KeywordMatcher] = None

        self.keyword_matcher = keyword_matcher or SetKeywordMatcher()

    def compute_metric(self, team_week_df):
        # Load data
        data_loader = DataLoader(self.data_path)
        df = data_loader.load_data()

        # Preprocess text
        df['cleaned_text'] = df['response_text'].apply(self.text_processor.preprocess)

        # Count keywords
        df['success_count'] = df['cleaned_text'].apply(
            lambda x: self.keyword_matcher.count_keywords(x, success_set)
        )
        df['failure_count'] = df['cleaned_text'].apply(
            lambda x: self.keyword_matcher.count_keywords(x, failure_set)
        )

        # Calculate scores
        df['text_length'] = df['cleaned_text'].apply(lambda x: len(x.split()))
        df['success_score'] = df['success_count'] / df['text_length']
        df['failure_score'] = df['failure_count'] / df['text_length']
        df['rule_based_score'] = df['success_score'] - df['failure_score']

        return df

---

class ConflictCalculator(TextMetric):
    def __init__(self, task_words, rel_words):
        self.task_words = task_words
        self.rel_words = rel_words

    def compute_metric(self, team_week_df):
        task_scores = []
        rel_scores = []
        for txt in team_week_df['text']:
            text_str = str(txt)
            if text_str.lower() == 'nan' or not text_str.strip():
                task_scores.append(np.nan)
                rel_scores.append(np.nan)
                continue
            t = text_str.lower()
            task = sum(t.count(w) for w in self.task_words)
            rel = sum(t.count(w) for w in self.rel_words)
            task_scores.append(task)
            rel_scores.append(rel)
        return np.nanmean(task_scores), np.nanmean(rel_scores)

class ParticipationCalculator(TextMetric):
    def compute_metric(self, team_week_df):
        if team_week_df['student_id'].nunique() < 3:
            return np.nan

        valid_texts = team_week_df['text'].apply(lambda x: str(x) if str(x).strip().lower() != 'nan' else np.nan).dropna()
        counts = valid_texts.astype(str).str.split().str.len().values

        if len(counts) == 0 or np.nansum(counts) == 0:
            return np.nan

        if len(counts) <= 1:
            return np.nan

        diffs = np.abs(np.subtract.outer(counts, counts))
        mean_counts = np.nanmean(counts)

        if mean_counts == 0:
            return np.nan

        gini = diffs.sum() / (2 * len(counts)**2 * mean_counts)
        return 1 - gini

class NoveltyCalculator(TextMetric):
    def compute_metric(self, team_week_df):
        texts = team_week_df['text'].tolist()

        if len(texts) < 2:
            return 0.0

        non_empty_texts = [str(text) for text in texts if str(text).strip() and str(text).lower() != 'nan']

        if len(non_empty_texts) < 2:
            return 0.0

        try:
            tf = TfidfVectorizer(min_df=1, token_pattern=r'(?u)\b\w+\b').fit_transform(non_empty_texts)

            if tf.shape[0] < 2 or tf.shape[1] == 0:
                return 0.0

            dist = cosine_distances(tf)
            np.fill_diagonal(dist, np.nan)
            return np.nanmean(dist)
        except ValueError as e:
            if "empty vocabulary" in str(e):
                return 0.0
            raise

class CognitiveCalculator(TextMetric):
    def __init__(self, insight_words, causation_words):
        self.insight_words = insight_words
        self.causation_words = causation_words

    def compute_metric(self, team_week_df):
        ins_scores = []
        cau_scores = []
        for txt in team_week_df['text']:
            text_str = str(txt)
            if text_str.lower() == 'nan' or not text_str.strip():
                ins_scores.append(np.nan)
                cau_scores.append(np.nan)
                continue
            t = text_str.lower()
            ins = sum(t.count(w) for w in self.insight_words)
            cau = sum(t.count(w) for w in self.causation_words)
            ins_scores.append(ins)
            cau_scores.append(cau)
        return np.nanmean(ins_scores), np.nanmean(cau_scores)


# Calculate individual LSM features per response
class IndividualLSMFeaturesCalculator:
    def __init__(self, fw_dict):
        self.fw_dict = fw_dict

    def _count_fw_ratio(self, text, lexicon):
        text_str = str(text)
        if text_str.lower() == 'nan' or not text_str.strip():
            return np.nan
        words = text_str.lower().split()
        if not words:
            return np.nan
        return sum(1 for w in words if w in lexicon) / len(words)

    def compute_individual_features(self, long_df):
        modified_long_df = long_df.copy()
        lsm_fw_cols = []
        for cat, lexicon in self.fw_dict.items():
            col_name = f'lsm_fw_{cat}'
            modified_long_df[col_name] = modified_long_df['text'].apply(lambda x: self._count_fw_ratio(x, lexicon))
            lsm_fw_cols.append(col_name)

        if lsm_fw_cols:
            modified_long_df['individual_lsm_composite'] = modified_long_df[lsm_fw_cols].mean(axis=1, skipna=True)
        else:
            modified_long_df['individual_lsm_composite'] = np.nan

        return modified_long_df


class MetricsProcessor:
    def __init__(self, long_df, metric_calculators):
        self.long_df = long_df
        self.metric_calculators = metric_calculators

    def process_metrics(self):
        metric_rows = []
        for (team, week), df_tw in self.long_df.groupby(['team_id', 'week'], dropna=False):
            if df_tw.empty:
                continue

            row = {'team_id': team, 'week': week}
            for metric_name, calculator in self.metric_calculators.items():
                # The individual calculators are responsible for copying if they modify df_tw.
                # Here, df_tw is already a subset, so we pass it directly to avoid an extra copy
                # unless the calculator specifically needs to modify it (which it now does internally
                # with its own copy if needed).
                if metric_name == 'conflict':
                    task_conflict, relationship_conflict = calculator.compute_metric(df_tw)
                    row['task_conflict'] = task_conflict
                    row['relationship_conflict'] = relationship_conflict
                elif metric_name == 'causation':
                    insight, causation = calculator.compute_metric(df_tw)
                    row['insight'] = insight
                    row['causation'] = causation
                else:
                    row[metric_name] = calculator.compute_metric(df_tw)
            metric_rows.append(row)

        return pd.DataFrame(metric_rows)


class ExcelExporter(Exporter):
    def __init__(self, output_folder):
        self.output_folder = output_folder

    def export(self, df_to_export, filename_suffix="", **kwargs):
        if df_to_export.empty:
            print("DataFrame for Excel export is empty. No data to save.")
            return

        timenow_str = datetime.now().strftime('%Y%m%d%H%M')
        # output_filename = f'{_raw_semester_year_setting}_{filename_suffix}_{timenow_str}.xlsx'
        output_filename = f'Metrics_{filename_suffix}_{timenow_str}.xlsx'
        output_path = os.path.join(self.output_folder, output_filename)

        try:
            df_to_export.to_excel(output_path, index=False, engine='openpyxl', **kwargs)
            print(f"\nSaved Excel metrics to: {output_path}\n")
            # print(df_to_export.head())
        except (TypeError, IndexError) as e:
            print(f"\nError saving to Excel: {e}")
            print("This often indicates a version incompatibility between pandas and openpyxl.")
        except Exception as e:
            print(f"\nAn unexpected error occurred while saving to Excel: {e}")


class CSVExporter(Exporter):
    def __init__(self, output_folder):
        self.output_folder = output_folder

    def export(self, df_to_export, filename_suffix="", **kwargs):
        if df_to_export.empty:
            print("DataFrame for CSV export is empty. No data to save.")
            return

        timenow_str = datetime.now().strftime('%Y%m%d%H%M')
        output_filename = f'Metrics_{filename_suffix}_{timenow_str}.csv.gz'
        output_path = os.path.join(self.output_folder, output_filename)

        try:
            df_to_export.to_csv(output_path, index=False, compression='gzip', **kwargs)
            print(f"\nSaved compressed metrics to: {output_path}\n")
            # print(df_to_export.head())
        except Exception as e:
            print(f"\nAn unexpected error occurred while saving to CSV: {e}")


class ParticipationEqualityPlotter(Plotter):
    def __init__(self, year_context='All'):
        self.year_context = year_context

    def plot(self, data_df, output_folder):
        # Use the centralized utility method for parsing team_id components
        data_df['cohort'], data_df['year'] = TeamUtils.parse_team_id_components(data_df['team_id'])

        years_to_plot = data_df['year'].unique()
        years_to_plot = [y for y in years_to_plot if pd.notna(y)]

        if not years_to_plot:
            print("No valid years found in the data for Participation Equality Plotting. Skipping.")
            return

        if self.year_context == 'All':
            filename_suffix_year = '_AllYears'
        else:
            filename_suffix_year = '_' + '_'.join(str(y) for y in self.year_context)
        timenow_str_plot = datetime.now().strftime('%Y%m%d%H%M')
        plot_output_path = os.path.join(output_folder, f'participation_cohorts_mean_std{filename_suffix_year}_{timenow_str_plot}.pdf')

        all_plotted_means = []
        all_plotted_stds = []

        for year in years_to_plot:
            year_df = data_df[data_df['year'] == year].copy()
            cohort_weekly_participation_mean_year = year_df.groupby(['cohort', 'week'])['participation_equality'].mean().reset_index()
            cohort_weekly_participation_std_year = year_df.groupby(['cohort', 'week'])['participation_equality'].std().reset_index()

            if not cohort_weekly_participation_mean_year.empty:
                all_plotted_means.append(cohort_weekly_participation_mean_year['participation_equality'])
            if not cohort_weekly_participation_std_year.empty:
                all_plotted_stds.append(cohort_weekly_participation_std_year['participation_equality'])

        global_min_val = np.inf
        global_max_val = -np.inf

        if all_plotted_means or all_plotted_stds:
            combined_plotted_values = pd.concat(all_plotted_means + all_plotted_stds).dropna()
            if not combined_plotted_values.empty:
                global_min_val = combined_plotted_values.min()
                global_max_val = combined_plotted_values.max()

        # Apply a buffer to the global min and max values
        buffer_percentage = 0.1 # 5% buffer
        if global_min_val == np.inf: # No data to plot, use a default reasonable range
            global_pe_min = -0.1 # Default reasonable min if no data
        else:
            global_pe_min = global_min_val * (1 + buffer_percentage) if global_min_val < 0 else global_min_val * (1 - buffer_percentage)

        if global_max_val == -np.inf:
            global_pe_max = 1.1
        else:
            global_pe_max = global_max_val * (1 + buffer_percentage) if global_max_val > 0 else global_max_val * (1 - buffer_percentage)

        if global_pe_min >= global_pe_max:
            global_pe_min = -0.1
            global_pe_max = 1.1

        with PdfPages(plot_output_path) as pdf:
            for year in years_to_plot:
                year_df = data_df[data_df['year'] == year].copy()
                cohort_weekly_participation_mean_year = year_df.groupby(['cohort', 'week'])['participation_equality'].mean().reset_index()
                cohort_weekly_participation_std_year = year_df.groupby(['cohort', 'week'])['participation_equality'].std().reset_index()

                if not cohort_weekly_participation_mean_year.empty:
                    fig, ax = plt.subplots(figsize=(12, 7)) # Get figure and axes objects
                    # Get the default color cycle
                    prop_cycle = plt.rcParams['axes.prop_cycle']
                    colors = prop_cycle.by_key()['color']

                    for i, (cohort_name, group_mean) in enumerate(cohort_weekly_participation_mean_year.groupby('cohort')):
                        color = colors[i % len(colors)] # Cycle through colors
                        # Plot mean
                        ax.plot(group_mean['week'], group_mean['participation_equality'], marker='o', linestyle='-', color=color, label=f'{cohort_name} (Mean)')

                        # Find corresponding std data
                        group_std = cohort_weekly_participation_std_year[cohort_weekly_participation_std_year['cohort'] == cohort_name]
                        ax.plot(group_std['week'], group_std['participation_equality'], marker='x', linestyle='--', color=color, label=f'{cohort_name} (Std Dev)')

                    # Add titles and labels
                    ax.set_title(f'Weekly Mean and Std Dev of Participation Equality by Cohort - Year {year}')
                    ax.set_xlabel('Week')
                    ax.set_ylabel('Participation Equality')
                    ax.legend(title='Cohort Metric', bbox_to_anchor=(1.05, 1), loc='upper left')
                    ax.grid(True)
                    # Ensure x-ticks are only for weeks present in the current year's data
                    ax.set_xticks(range(1, 13)) # Always show 12 weeks
                    ax.set_ylim(global_pe_min, global_pe_max) # Set consistent y-axis limits

                    plt.tight_layout()
                    pdf.savefig(fig) # Save the current figure to the PDF
                    plt.close(fig)   # Close the figure to free up memory
                    # print(f"  Added participation plot for year {year} to consolidated PDF.")
                else:
                    print(f"  No data for year {year} to plot participation equality.")
        print(f"Saved consolidated participation plots to: {plot_output_path}")

# New Plotter implementation for LSM Trends
class LSMTrendsPlotter(Plotter):
    def __init__(self, year_context='All'):
        self.year_context = year_context

    def plot(self, data_df, output_folder):
        # Use the centralized utility method for parsing team_id components
        data_df['cohort'], data_df['year'] = TeamUtils.parse_team_id_components(data_df['team_id'])

        years_to_plot = data_df['year'].unique()
        years_to_plot = [y for y in years_to_plot if pd.notna(y)]

        if not years_to_plot:
            print("No valid years found in the data for LSM Trends Plotting. Skipping.")
            return

        if self.year_context == 'All':
            filename_suffix_year = '_AllYears'
        else:
            filename_suffix_year = '_' + '_'.join(str(y) for y in self.year_context)
        timenow_str_plot = datetime.now().strftime('%Y%m%d%H%M')
        plot_output_path = os.path.join(output_folder, f'lsm_cohorts_mean_std{filename_suffix_year}_{timenow_str_plot}.pdf')

        all_plotted_means = []
        all_plotted_stds = []

        for year in years_to_plot:
            year_df = data_df[data_df['year'] == year].copy()
            cohort_weekly_lsm_mean = year_df.groupby(['cohort', 'week'])['lsm'].mean().reset_index()
            cohort_weekly_lsm_std = year_df.groupby(['cohort', 'week'])['lsm'].std().reset_index()

            if not cohort_weekly_lsm_mean.empty:
                all_plotted_means.append(cohort_weekly_lsm_mean['lsm'])
            if not cohort_weekly_lsm_std.empty:
                all_plotted_stds.append(cohort_weekly_lsm_std['lsm'])

        global_min_val = np.inf
        global_max_val = -np.inf

        if all_plotted_means or all_plotted_stds:
            combined_plotted_values = pd.concat(all_plotted_means + all_plotted_stds).dropna()
            if not combined_plotted_values.empty:
                global_min_val = combined_plotted_values.min()
                global_max_val = combined_plotted_values.max()

        buffer_percentage = 0.05
        if global_min_val == np.inf:
            global_lsm_min = -0.3
        else:
            global_lsm_min = global_min_val * (1 + buffer_percentage) if global_min_val < 0 else global_min_val * (1 - buffer_percentage)

        if global_max_val == -np.inf:
            global_lsm_max = 1.0
        else:
            global_lsm_max = global_max_val * (1 + buffer_percentage) if global_max_val > 0 else global_max_val * (1 - buffer_percentage)

        if global_lsm_min >= global_lsm_max:
            global_lsm_min = -0.3
            global_lsm_max = 1.0

        with PdfPages(plot_output_path) as pdf:
            for year in years_to_plot:
                year_df = data_df[data_df['year'] == year].copy()
                cohort_weekly_lsm_mean = year_df.groupby(['cohort', 'week'])['lsm'].mean().reset_index()
                cohort_weekly_lsm_std = year_df.groupby(['cohort', 'week'])['lsm'].std().reset_index()

                if not cohort_weekly_lsm_mean.empty:
                    fig, ax = plt.subplots(figsize=(12, 7))
                    prop_cycle = plt.rcParams['axes.prop_cycle']
                    colors = prop_cycle.by_key()['color']

                    for i, (cohort_name, group_mean) in enumerate(cohort_weekly_lsm_mean.groupby('cohort')):
                        color = colors[i % len(colors)]
                        ax.plot(group_mean['week'], group_mean['lsm'], marker='o', linestyle='-', color=color, label=f'{cohort_name} (Mean)')
                        group_std = cohort_weekly_lsm_std[cohort_weekly_lsm_std['cohort'] == cohort_name]
                        ax.plot(group_std['week'], group_std['lsm'], marker='x', linestyle='--', color=color, label=f'{cohort_name} (Std Dev)')

                    ax.set_title(f'Weekly Mean and Std Dev of LSM by Cohort - Year {year}')
                    ax.set_xlabel('Week')
                    ax.set_ylabel('LSM Score')
                    ax.legend(title='Cohort Metric', bbox_to_anchor=(1.05, 1), loc='upper left')
                    ax.grid(True)
                    ax.set_xticks(range(1, 13))
                    ax.set_ylim(global_lsm_min, global_lsm_max)

                    plt.tight_layout()
                    pdf.savefig(fig)
                    plt.close(fig)
                else:
                    print(f"  No LSM data for year {year} to plot.")
        print(f"Saved consolidated LSM plots to: {plot_output_path}")


class IndividualLSMWeekValidation(ValidationCheck):
    def __init__(self, wide_individual_lsm_df):
        self.df = wide_individual_lsm_df

    def validate(self):
        print("\n--- Check: Valid Individual LSM Score Presence (Weeks 1 & 2) ---")

        col_week1 = 'individual_lsm_composite_week_1'
        col_week2 = 'individual_lsm_composite_week_2'

        has_week1_score = self.df[col_week1].notna() if col_week1 in self.df.columns else pd.Series(False, index=self.df.index)
        has_week2_score = self.df[col_week2].notna() if col_week2 in self.df.columns else pd.Series(False, index=self.df.index)

        valid_week1_only = self.df[has_week1_score & ~has_week2_score].shape[0]
        valid_week2_only = self.df[~has_week1_score & has_week2_score].shape[0]
        valid_neither_week = self.df[~has_week1_score & ~has_week2_score].shape[0]
        valid_both_weeks = self.df[has_week1_score & has_week2_score].shape[0]

        validation_results = pd.DataFrame({
            'Category': ['Valid in Week 1 Only', 'Valid in Week 2 Only', 'Valid Neither Week 1 Nor 2', 'Valid in Both Weeks'],
            'Number of Individuals': [valid_week1_only, valid_week2_only, valid_neither_week, valid_both_weeks]
        })

        print(validation_results.to_string()) # Changed to print.to_string() for robustness
        print("------------------------------------------------------------------\n")

class ScoreValueValidation(ValidationCheck):
    def __init__(self, metrics_df):
        self.df = metrics_df

    def validate(self):
        print("\n--- Check: Scores exactly equal to 1.0 for LSM or Participation Equality ---")

        problematic_entries = self.df[
            (self.df['lsm'] == 1.0) | (self.df['participation_equality'] == 1.0)
        ]

        if not problematic_entries.empty:
            print(f"Found {len(problematic_entries)} team-week entries where LSM or Participation Equality is exactly 1.0.")
            print("First 5 problematic entries:")
            display(problematic_entries.head())
        else:
            print("No team-week entries found where LSM or Participation Equality is exactly 1.0.")
        print("------------------------------------------------------------------\n")


class ExtremeScoreTextAnalyzer:
    def __init__(self, long_df, extreme_score_entries):
        self.long_df = long_df
        self.extreme_score_entries = extreme_score_entries

    def analyze_extreme_scores(self):
        print("\n--- Analyzing Raw Text Data for Extreme Scores ---")

        analysis_results = []

        for index, row in self.extreme_score_entries.iterrows():
            team_id = row['team_id']
            week = row['week']

            filtered_texts_df = self.long_df[
                (self.long_df['team_id'] == team_id) &
                (self.long_df['week'] == week)
            ].copy()

            if filtered_texts_df.empty:
                analysis_results.append({
                    'team_id': team_id,
                    'week': week,
                    'num_contributors': 0,
                    'total_responses': 0,
                    'avg_response_length': np.nan,
                    'num_unique_texts': 0,
                    'has_identical_texts': False,
                    'notes': 'No text entries found for this team-week.'
                })
                continue

            num_contributors = filtered_texts_df['student_id'].nunique()
            total_responses = len(filtered_texts_df)

            response_lengths = filtered_texts_df['text'].apply(lambda x: len(str(x).split()))
            avg_response_length = response_lengths.mean() if not response_lengths.empty else np.nan

            num_unique_texts = filtered_texts_df['text'].nunique()
            has_identical_texts = total_responses > num_unique_texts

            analysis_results.append({
                'team_id': team_id,
                'week': week,
                'num_contributors': num_contributors,
                'total_responses': total_responses,
                'avg_response_length': avg_response_length,
                'num_unique_texts': num_unique_texts,
                'has_identical_texts': has_identical_texts
            })

        results_df = pd.DataFrame(analysis_results)

        print(f"Analyzed {len(results_df)} extreme team-week entries.")
        print("Preview of analysis results (first 5 rows):")
        display(results_df.head())
        print("----------------------------------------------------------\n")

        return results_df

class NonTextResponseValidator(ValidationCheck):
    def __init__(self, long_df, exporter):
        self.long_df = long_df
        self.exporter = exporter

    def validate(self):
        print("\n--- Validating and Cleaning Non-Text Responses ---")

        original_rows = len(self.long_df)
        numeric_mask = self.long_df['text'].astype(str).str.fullmatch(r'^\s*-?\d+(\.\d+)?\s*$')
        letter_mask = self.long_df['text'].astype(str).str.fullmatch(r'^\s*[a-zA-Z]\s*$')
        combined_mask = numeric_mask | letter_mask

        initial_matches = combined_mask.sum()

        if initial_matches > 0:
            problematic_raw_entries = self.long_df[combined_mask].copy()
            self.exporter.export(problematic_raw_entries, filename_suffix='_Raw_NonText_Responses_Before_Cleaning')
            self.long_df.loc[combined_mask, 'text'] = np.nan

            print(f"Identified and replaced {initial_matches} non-text responses (single numeric or single letter) with np.nan.")

        else:
            print("No non-text responses (single numeric or single letter) found. No modifications made.")

        print(f"Original long_df entries: {original_rows}")
        print(f"Current long_df entries after cleaning (rows are not dropped): {len(self.long_df)}")

        non_string_non_nan_mask = self.long_df['text'].apply(lambda x: pd.notna(x) and not isinstance(x, str))
        non_string_non_nan_count = non_string_non_nan_mask.sum()

        if non_string_non_nan_count > 0:
            print(f"Warning: Found {non_string_non_nan_count} non-NaN entries in 'text' column that are not strings AFTER cleaning.")
            display(self.long_df[non_string_non_nan_mask].head())
        else:
            print("All remaining non-NaN entries in 'text' column are strings.")

        print("------------------------------------------------------------------\n")


# Renamed from PersonalDescriptionAnalyzer to IndividualResponseAnalyzer
class IndividualResponseAnalyzer:
    def __init__(self, long_df, lexicon_loader, excel_exporter, analysis_targets=None):
        self.long_df = long_df
        self.lexicon_loader = lexicon_loader
        self.excel_exporter = excel_exporter
        # analysis_targets is expected to be a dictionary: {'variable_name': week_number}
        self.analysis_targets = analysis_targets if analysis_targets is not None else {'personal_desc': 1}

    def analyze_and_export(self):
        print(f"\n--- Analyzing individual entries for {self.analysis_targets} ---")

        base_id_cols = ['student_id', 'team_id', 'semester_id']
        # Collect all unique student_ids that appear in any of the target variable/week combinations
        all_relevant_students = pd.DataFrame(columns=base_id_cols)
        for var_name, week in self.analysis_targets.items():
            students_for_target = self.long_df[
                (self.long_df['variable_name'] == var_name) &
                (self.long_df['week'] == week)
            ][base_id_cols].drop_duplicates()
            all_relevant_students = pd.concat([all_relevant_students, students_for_target]).drop_duplicates().reset_index(drop=True)

        if all_relevant_students.empty:
            print(f"No relevant student entries found for any of the analysis targets {self.analysis_targets}. Skipping individual-level metric calculations.")
            return pd.DataFrame() # Return empty DataFrame

        final_individual_metrics_df = all_relevant_students.copy()

        # Instantiate individual metric calculators (once outside the inner loop)
        individual_lsm_calculator = IndividualLSMFeaturesCalculator(self.lexicon_loader.fw_dict)
        identity_calculator = IdentityCalculator()
        safety_calculator = SafetyCalculator(self.lexicon_loader.hedges, self.lexicon_loader.apologies)
        conflict_calculator = ConflictCalculator(self.lexicon_loader.task_words, self.lexicon_loader.rel_words)
        cognitive_calculator = CognitiveCalculator(self.lexicon_loader.insight_words, self.lexicon_loader.causation_words)

        for var_name, week_num in self.analysis_targets.items():
            temp_df = self.long_df[
                (self.long_df['variable_name'] == var_name) &
                (self.long_df['week'] == week_num)
            ].copy()

            # Text cleaning for this specific variable's responses
            temp_df = temp_df.dropna(subset=['text']).copy()
            temp_df['text'] = temp_df['text'].astype(str).str.strip()
            temp_df = temp_df[temp_df['text'] != ''].copy()

            if temp_df.empty:
                print(f"  No valid entries for '{var_name}' for Week {week_num} after cleaning.")
                # Create a placeholder row of NaNs for these students so that merge doesn't drop them
                placeholder_df = all_relevant_students.copy()
                placeholder_df = placeholder_df[~placeholder_df['student_id'].isin(temp_df['student_id'])] # Students that didn't contribute to this var/week
                for col_suffix in ['text', 'individual_lsm_composite', 'identity_we_i', 'safety', 'task_conflict', 'relationship_conflict', 'insight', 'causation']:
                    placeholder_df[f'{var_name}_week_{week_num}_{col_suffix}'] = np.nan
                # Placeholder for lsm_fw_ columns as well
                for fw_cat in self.lexicon_loader.fw_dict.keys():
                    placeholder_df[f'{var_name}_week_{week_num}_lsm_fw_{fw_cat}'] = np.nan

                # Ensure placeholder_df only contains columns intended for merging
                cols_to_merge_from_placeholder = [col for col in placeholder_df.columns if col not in base_id_cols]
                final_individual_metrics_df = pd.merge(final_individual_metrics_df, placeholder_df[base_id_cols + cols_to_merge_from_placeholder], on=base_id_cols, how='left')
                continue

            # Calculate metrics for the current variable name
            current_var_metrics_df = temp_df[base_id_cols + ['text']].copy() # Include original text

            # Calculate individual LSM features
            current_var_metrics_df = individual_lsm_calculator.compute_individual_features(current_var_metrics_df)

            # Apply other single-response metrics
            current_var_metrics_df['identity_we_i'] = current_var_metrics_df['text'].apply(lambda x: identity_calculator.compute_metric(pd.DataFrame({'text': [x]})))
            current_var_metrics_df['safety'] = current_var_metrics_df['text'].apply(lambda x: safety_calculator.compute_metric(pd.DataFrame({'text': [x]})))

            conflict_results = current_var_metrics_df['text'].apply(lambda x: conflict_calculator.compute_metric(pd.DataFrame({'text': [x]})))
            current_var_metrics_df['task_conflict'] = conflict_results.apply(lambda x: x[0])
            current_var_metrics_df['relationship_conflict'] = conflict_results.apply(lambda x: x[1])

            cognitive_results = current_var_metrics_df['text'].apply(lambda x: cognitive_calculator.compute_metric(pd.DataFrame({'text': [x]})))
            current_var_metrics_df['insight'] = cognitive_results.apply(lambda x: x[0])
            current_var_metrics_df['causation'] = cognitive_results.apply(lambda x: x[1])

            # Rename columns to include variable_name and week prefix
            metrics_to_rename = [
                'text', # Keep original text associated with this variable
                'individual_lsm_composite', 'identity_we_i', 'safety',
                'task_conflict', 'relationship_conflict', 'insight', 'causation'
            ]
            # Also include lsm_fw_ categories if present (from individual_lsm_calculator)
            lsm_fw_cols = [col for col in current_var_metrics_df.columns if col.startswith('lsm_fw_')]
            metrics_to_rename.extend(lsm_fw_cols)

            rename_map = {col: f'{var_name}_week_{week_num}_{col}' for col in metrics_to_rename if col in current_var_metrics_df.columns}
            current_var_metrics_df = current_var_metrics_df.rename(columns=rename_map)

            # Merge into final_individual_metrics_df
            final_individual_metrics_df = pd.merge(
                final_individual_metrics_df,
                current_var_metrics_df[[col for col in current_var_metrics_df.columns if col not in ['question', 'variable_name', 'week']]], # Exclude irrelevant cols for merge
                on=base_id_cols,
                how='left'
            )

        if final_individual_metrics_df.empty:
            print("No consolidated individual metrics found after processing all variable types.")
            return pd.DataFrame()

        print(f"Total unique students with consolidated individual metrics: {len(final_individual_metrics_df)}")
        print("Preview of consolidated individual metrics:")
        display(final_individual_metrics_df.head())

        # Export the results to a separate Excel file
        # filename_prefix = '_ConsolidatedIndividualMetrics_Week' + '_'.join(str(w) for w in self.analysis_targets.values()) + '_Vars_' + '_'.join(self.analysis_targets.keys())
        filename_prefix = '_ConsolidatedIndividualMetrics_Week'
        self.excel_exporter.export(final_individual_metrics_df, filename_suffix=filename_prefix)

        print("------------------------------------------------------------------\n")
        return final_individual_metrics_df


# -------------------------------------------------------------------
# Main Method for Data Analysis
# -------------------------------------------------------------------

def run_full_analysis(config):
    # Access configuration parameters from the config object
    CODEBOOK_FILE = config.CODEBOOK_FILE
    LMS_LEXICON_FILE = config.LMS_LEXICON_FILE
    LIWC_LEXICON_FILE = config.LIWC_LEXICON_FILE
    SUCC_FAIL_LEXICON_FILE = config.SUCC_FAIL_LEXICON_FILE
    input_data_dir = config.input_data_dir
    semester_year = config.semester_year
    OUTPUT_FOLDER = config.OUTPUT_FOLDER

    # Instantiate the ExcelDataLoader and load raw comments data using RawCommentsDataLoader
    excel_data_loader = ExcelDataLoader()
    raw_comments_data_loader = RawCommentsDataLoader(input_data_dir, semester_year, excel_data_loader)
    raw_comments_data_df = raw_comments_data_loader.load_raw_comments_data()

    # --- Load codebook data and instantiate the CodebookProcessor with pre-loaded DataFrames
    codebook_df = excel_data_loader.load_data(CODEBOOK_FILE)
    codebook_processor = CodebookProcessor(codebook_df, raw_comments_data_df)
    long_df = codebook_processor.process_codebook_and_data()

    # Instantiate Exporters
    excel_exporter = ExcelExporter(OUTPUT_FOLDER)
    csv_exporter = CSVExporter(OUTPUT_FOLDER)

    # --- Integrate NonTextResponseValidator ---
    non_text_validator = NonTextResponseValidator(long_df, excel_exporter)
    non_text_validator.validate()

    # long_df is now cleaned; proceed with specific analysis for personal descriptions
    # And then prepare long_df_for_metrics for team-level metrics

    # Instantiate LexiconLoader with all required file paths
    lexicon_loader = LexiconLoader(LMS_LEXICON_FILE, LIWC_LEXICON_FILE,SUCC_FAIL_LEXICON_FILE)
    lexicon_loader.display_fw_summary()

    individual_response_metrics_df = pd.DataFrame() # Initialize empty DataFrame

    # --- Analyze specified individual entries for specific weeks and export results (if enabled) ---
    if config.run_individual_analysis:
        analysis_targets = {
            'personal_desc': 1,
            'comment_at1': 1,
            'comment_at2': 2,
            'comment_at3': 3,
            'comment_at4': 4,
            'comment_at5': 5,
            'comment_at6': 6,
            'comment_at7': 7,
            'comment_at8': 8,
            'comment_at9': 9,
            'comment_at10': 10,
            'comment_at11': 11
        }
        individual_response_analyzer = IndividualResponseAnalyzer(
            long_df,
            lexicon_loader,
            excel_exporter,
            analysis_targets=analysis_targets
        )
        individual_response_metrics_df = individual_response_analyzer.analyze_and_export()


    # --- Proceed with team-level analysis (if enabled) ---
    if config.run_team_analysis:
        long_df_for_metrics = long_df.dropna(subset=['text', 'week']).copy()

        if long_df_for_metrics.empty:
            print("long_df_for_metrics is empty after cleaning. No metrics to calculate or save.")
        else:
            # --- Add cohort and year to long_df_for_metrics before individual LSM feature calculation ---
            # Centralized parsing using TeamUtils
            long_df_for_metrics['cohort'], long_df_for_metrics['year'] = TeamUtils.parse_team_id_components(long_df_for_metrics['team_id'])

            # Instantiate IndividualLSMFeaturesCalculator and enrich long_df_for_metrics
            individual_lsm_calculator = IndividualLSMFeaturesCalculator(lexicon_loader.fw_dict)
            long_df_for_metrics = individual_lsm_calculator.compute_individual_features(long_df_for_metrics)

            # --- Export individual LSM features in wide format ---
            lsm_feature_cols = [col for col in long_df_for_metrics.columns if col.startswith('lsm_fw_') or col == 'individual_lsm_composite']
            individual_id_cols = ['year', 'cohort', 'team_id', 'student_id', 'semester_id']
            individual_lsm_df_for_pivot = long_df_for_metrics[individual_id_cols + ['week'] + lsm_feature_cols].copy()
            wide_individual_lsm_df = individual_lsm_df_for_pivot.pivot_table(
                index=individual_id_cols,
                columns='week',
                values=lsm_feature_cols
            )

            # Flatten the multi-level column index dynamically based on its type
            new_wide_individual_lsm_columns = []
            if isinstance(wide_individual_lsm_df.columns, pd.MultiIndex):
                for col_tuple in wide_individual_lsm_df.columns:
                    new_wide_individual_lsm_columns.append(f'{col_tuple[0]}_week_{int(col_tuple[1])}')
            elif len(lsm_feature_cols) == 1:
                # If it's not a MultiIndex and there's only one feature, assume column names are just week numbers (floats)
                feature_name = lsm_feature_cols[0]
                for week_val in wide_individual_lsm_df.columns:
                    new_wide_individual_lsm_columns.append(f'{feature_name}_week_{int(week_val)}')
            else:
                # Fallback for unexpected single-level index type, just convert to string
                for col_name in wide_individual_lsm_df.columns:
                    new_wide_individual_lsm_columns.append(str(col_name))
            wide_individual_lsm_df.columns = new_wide_individual_lsm_columns

            wide_individual_lsm_df = wide_individual_lsm_df.reset_index()
            print("\nWide-format Individual LSM Features DataFrame preview:")
            display(wide_individual_lsm_df.head())

            # Export individual LSM features to CSV
            csv_exporter.export(wide_individual_lsm_df, filename_suffix='_Individual_LSM_Features')

            # --- Perform Validation Check for LSM scores in Week 1 and Week 2 ---
            lsm_week_validator = IndividualLSMWeekValidation(wide_individual_lsm_df)
            lsm_week_validator.validate()

            # Instantiate individual metric calculators
            metric_calculators = {
                'lsm': LSMCalculator(lexicon_loader.fw_dict),
                'coordination': CoordinationCalculator(),
                'participation_equality': ParticipationCalculator(),
                'novelty': NoveltyCalculator(),
                'identity_we_i': IdentityCalculator(),
                'safety': SafetyCalculator(lexicon_loader.hedges, lexicon_loader.apologies),
                'conflict': ConflictCalculator(lexicon_loader.task_words, lexicon_loader.rel_words),
                'causation': CognitiveCalculator(lexicon_loader.insight_words, lexicon_loader.causation_words)
            }

            # Instantiate MetricsProcessor and process metrics
            metrics_processor = MetricsProcessor(long_df_for_metrics, metric_calculators)
            calculated_metrics_df = metrics_processor.process_metrics()

            # After all calculations, create the full metrics_df including NaN entries
            all_team_weeks = long_df[['team_id', 'week']].drop_duplicates().copy()
            all_team_weeks['team_id'] = all_team_weeks['team_id'].astype(str)
            metrics_df = pd.merge(all_team_weeks, calculated_metrics_df, on=['team_id', 'week'], how='left')
            # Centralized parsing using TeamUtils
            metrics_df['cohort'], metrics_df['year'] = TeamUtils.parse_team_id_components(metrics_df['team_id'])

            if metrics_df.empty:
                print("metrics_df is empty. No data to calculate or save.")
            else:
                participation_plotter = ParticipationEqualityPlotter(year_context=semester_year)
                participation_plotter.plot(metrics_df.copy(), OUTPUT_FOLDER)

                lsm_trends_plotter = LSMTrendsPlotter(year_context=semester_year)
                lsm_trends_plotter.plot(metrics_df.copy(), OUTPUT_FOLDER)

                lsm_cols = ['year', 'cohort', 'team_id', 'week', 'lsm']
                lsm_metrics_df = metrics_df[lsm_cols].copy()
                wide_lsm_metrics_df = lsm_metrics_df.pivot_table(
                    index=['year', 'cohort', 'team_id'],
                    columns='week',
                    values='lsm'
                )
                wide_lsm_metrics_df.columns = [f'lsm_week_{int(col)}' for col in wide_lsm_metrics_df.columns]
                wide_lsm_metrics_df = wide_lsm_metrics_df.reset_index()

                print("\nWide-format LSM metrics DataFrame preview:\n")
                display(wide_lsm_metrics_df.head())

                team_lsm_week_validator = TeamLSMWeekValidation(wide_lsm_metrics_df)
                team_lsm_week_validator.validate()

                lsm_analyzer = LSMAnalyzer(wide_lsm_metrics_df)
                lsm_analyzer.analyze_lsm_trends()

                excel_exporter.export(wide_lsm_metrics_df, filename_suffix='_LSM')

                all_metric_cols = [col for col in metrics_df.columns if col not in ['team_id', 'week', 'cohort', 'year']]
                wide_metrics_df = metrics_df.pivot_table(
                    index=['year', 'cohort', 'team_id'],
                    columns='week',
                    values=all_metric_cols
                )
                wide_metrics_df.columns = [f'{col[0]}_week_{int(col[1])}' if isinstance(col, tuple) else col for col in wide_metrics_df.columns.values]
                wide_metrics_df = wide_metrics_df.reset_index()
                print("\nWide-format all metrics DataFrame Complete:\n")
                # display(wide_metrics_df.head())

                correlation_analyzer = CorrelationAnalyzer(wide_metrics_df)
                correlation_analyzer.analyze_lsm_correlations()

                # Export the correlation results to a separate Excel file
                correlations_to_export = correlation_analyzer.get_all_correlations_df()
                if not correlations_to_export.empty:
                    excel_exporter.export(correlations_to_export, filename_suffix='_WeeklyLSMCorrelations')
                else:
                    print("No weekly LSM correlations to export.")

                excel_exporter.export(wide_metrics_df, filename_suffix='_AllMetrics')

                score_value_validator = ScoreValueValidation(metrics_df)
                score_value_validator.validate()

                extreme_scores_df_recalculated = metrics_df[
                    (metrics_df['lsm'] == 1.0) | (metrics_df['participation_equality'] == 1.0)
                ].copy()
                extreme_score_entries_recalculated = extreme_scores_df_recalculated[['team_id', 'week']].drop_duplicates().reset_index(drop=True)

                if not extreme_score_entries_recalculated.empty:
                    extreme_text_analyzer = ExtremeScoreTextAnalyzer(long_df, extreme_score_entries_recalculated)
                    extreme_scores_analysis_df = extreme_text_analyzer.analyze_extreme_scores()
                    excel_exporter.export(extreme_scores_analysis_df, filename_suffix='_ExtremeScoresTextAnalysis')
                else:
                    print("No extreme score entries found after re-calculating metrics. Skipping ExtremeScoreTextAnalyzer.")


# -------------------------------------------------------------------
# Container Process for X-Culture Data Analysis
# -------------------------------------------------------------------
# --- Check filesystem access ---
drive_checker = OSDriveCheck()
drive_checker.check_and_mount_drive()

# --- Show date and time at the start of the overall process ---
main_start_time = datetime.now()
print(f"Container process started at: {main_start_time}")

# --- Fetch Global Variables at the start of the overall process ---
config = Config()

if config.REDIRECT_WARNINGS_TO_FILE:
    print(f"Redirecting warnings to: {config.warnings_log_file}")
    with redirect_stderr_to_file(config.warnings_log_file):
        # Capture the results returned by run_full_analysis
        _results = run_full_analysis(config)
        # Assign global variables from the returned tuple, ensuring they are assigned.
        # Only assign if _results is a tuple and has the expected number of elements.
        # This needs careful handling if run_full_analysis becomes more complex.
        # For now, let's keep the global assignment logic inside run_full_analysis itself for simplicity as previously planned.
        pass # Actual assignment now happens inside run_full_analysis after global declaration
else:
    _results = run_full_analysis(config)
    pass # Actual assignment now happens inside run_full_analysis after global declaration

main_end_time = datetime.now()
duration = main_end_time - main_start_time
print(f"Container process finished at {main_start_time}, Duration {duration}")
#
# -----------------------------------------------------------------
# End of Container Process for X-Culture Data Analysis
# -----------------------------------------------------------------
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import nltk
import re
import json
import os
import yake
import spacy
import numpy as np
from datetime import datetime
from textblob import TextBlob
from matplotlib.backends.backend_pdf import PdfPages
from google.colab import drive
from collections import Counter, defaultdict
from nltk.corpus import stopwords
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from abc import ABC, abstractmethod

# Global VADER Analyzer instance for reuse
vader_analyzer_instance = SentimentIntensityAnalyzer()

# Interface for Text Analysis Components
class ITextAnalyzer(ABC):
    """Abstract Base Class for text analysis components."""
    @abstractmethod
    def analyze(self, comments_df: pd.DataFrame, var_week_map: dict, target_year: int, target_top_noun_phrases: int, results_container: dict):
        """Performs text analysis and stores results in the results_container."""
        pass

# Interface for Plotting Strategies
class IPlottingStrategy(ABC):
    """Abstract Base Class for plotting strategies."""
    @abstractmethod
    def plot(self, pdf_pages, data: dict):
        """Generates a specific plot and saves it to pdf_pages."""
        pass

# Interface for Clustering Strategies
class IClusteringStrategy(ABC):
    """Abstract Base Class for clustering strategies."""
    @abstractmethod
    def perform_clustering(self, data_df: pd.DataFrame, week1_sentiment_col: str, weekX_sentiment_col: str, target_k_clusters: int, k_range_optimizer: range, var_week_map: dict = None, cohort_sentiment_data: pd.DataFrame = None) -> dict:
        """Performs clustering and returns clustering results."""
        pass


class XCultureNounPhraseExtractor(ITextAnalyzer):
    def __init__(self):
        print(f"Initialized: Noun Phrase Extractor.")

    def analyze(self, comments_df: pd.DataFrame, var_week_map: dict, target_year: int, target_top_noun_phrases: int, results_container: dict):
        if comments_df.empty:
            print("comments_df is empty, skipping noun phrase extraction.")
            results_container.setdefault('yearly_top_noun_phrases', {})[target_year] = []
            return

        all_noun_phrases = []
        comment_columns = [col for col in var_week_map.keys() if col in comments_df.columns]

        for col in comment_columns:
            for text in comments_df[col].dropna():
                if isinstance(text, str) and text.strip():
                    blob = TextBlob(text)
                    all_noun_phrases.extend([np.lower() for np in blob.noun_phrases])

        if all_noun_phrases:
            noun_phrase_counts = Counter(all_noun_phrases)
            top_noun_phrases = noun_phrase_counts.most_common(target_top_noun_phrases)
            results_container.setdefault('yearly_top_noun_phrases', {})[target_year] = top_noun_phrases
        else:
            print(f"No noun phrases found for year {target_year}.")
            results_container.setdefault('yearly_top_noun_phrases', {})[target_year] = []


class XCultureYAKEKeywordExtractor(ITextAnalyzer):
    def __init__(self):
        print(f"Initialized: YAKE Keyword Extractor.")

    def analyze(self, comments_df: pd.DataFrame, var_week_map: dict, target_year: int, target_top_noun_phrases: int, results_container: dict):
        if comments_df.empty:
            print("comments_df is empty, skipping YAKE keyword extraction.")
            results_container.setdefault('yearly_yake_keywords', {})[target_year] = []
            return

        all_comments_text = []
        comment_columns = [col for col in var_week_map.keys() if col in comments_df.columns]

        for col in comment_columns:
            for text in comments_df[col].dropna():
                if isinstance(text, str) and text.strip():
                    all_comments_text.append(text)

        if all_comments_text:
            combined_text = " ".join(all_comments_text)
            kw_extractor = yake.KeywordExtractor(lan="en", n=3, dedupLim=0.9, top=target_top_noun_phrases, features=None)
            keywords = kw_extractor.extract_keywords(combined_text)
            keywords_sorted = sorted(keywords, key=lambda x: x[1])
            results_container.setdefault('yearly_yake_keywords', {})[target_year] = keywords_sorted
        else:
            results_container.setdefault('yearly_yake_keywords', {})[target_year] = []


class XCultureNationalityMiner(ITextAnalyzer):
    def __init__(self, geo_json_file):
        self.geo_json_file = geo_json_file
        self.country_data = self._load_country_data()

        self.name_to_country_map = {}
        self.geo_term_to_country_map = {}
        self.nationality_to_country_map = {}
        self._build_lookup_maps()
        self.country_names = sorted(list(self.name_to_country_map.keys()), key=len, reverse=True)
        self.geo_terms = sorted(list(self.geo_term_to_country_map.keys()), key=len, reverse=True)
        self.compiled_country_patterns = [(re.compile(r'\b' + re.escape(n) + r'\b', re.I), self.name_to_country_map[n]) for n in self.country_names]
        self.compiled_geo_patterns = [(re.compile(r'\b' + re.escape(t) + r'\b', re.I), self.geo_term_to_country_map[t]) for t in self.geo_terms]
        self.compiled_nat_patterns = [(re.compile(r'\b' + re.escape(nat) + r'\b', re.I), self.nationality_to_country_map[nat]) for nat in self.nationality_to_country_map]
        self.nlp = spacy.load("en_core_web_lg")
        print(f"Initialized: Optimized Batch Nationality Miner.")

    def _load_country_data(self):
        with open(self.geo_json_file, 'r') as f:
            return json.load(f)

    def _build_lookup_maps(self):
        for c in self.country_data['countries']:
            cname = c['name']
            self.name_to_country_map[cname.lower()] = cname
            for alias in c.get('aliases', []):
                self.name_to_country_map[alias.lower()] = cname
            if isinstance(c.get('capital'), str):
                self.geo_term_to_country_map[c['capital'].lower()] = cname
            for city in c.get('major_cities', []):
                self.geo_term_to_country_map[city.lower()] = cname
            for p in c.get('provinces', []):
                if 'province_name' in p: self.geo_term_to_country_map[p['province_name'].lower()] = cname
            for s in c.get('states', []):
                if 'state_name' in s: self.geo_term_to_country_map[s['state_name'].lower()] = cname
            nats = c.get('nationality', [])
            if isinstance(nats, str): nats = [nats]
            for n in nats: self.nationality_to_country_map[n.lower()] = cname

    def analyze(self, comments_df, var_week_map, target_year, target_top_noun_phrases, results_container):
        if 'personal_desc' not in comments_df.columns: return

        texts = comments_df['personal_desc'].astype(str).tolist()
        ner_results = [None] * len(texts)

        # 1. Batch process with NER
        # We disable unnecessary pipeline components to speed up processing
        for i, doc in enumerate(self.nlp.pipe(texts, disable=["tok2vec", "tagger", "parser", "attribute_ruler", "lemmatizer"], batch_size=500)):
            for ent in doc.ents:
                if ent.label_ == "GPE" and ent.text.lower() in self.name_to_country_map:
                    ner_results[i] = self.name_to_country_map[ent.text.lower()]
                    break

        final_results = []
        for i, text in enumerate(texts):
            if len(text.strip()) < 5:
                final_results.append('EMPTY')
                continue

            found = ner_results[i]
            text_l = text.lower()

            # 2. Sequential fallback Regex Matching if NER failed
            if not found:
                for pattern, cname in self.compiled_country_patterns:
                    if pattern.search(text_l): found = cname; break
            if not found:
                for pattern, cname in self.compiled_geo_patterns:
                    if pattern.search(text_l): found = cname; break
            if not found:
                for pattern, cname in self.compiled_nat_patterns:
                    if pattern.search(text_l): found = cname; break

            final_results.append(found if found else 'UNKNOWN')

        comments_df['nationality'] = final_results
        results_container.setdefault('yearly_nationality_results', {})[target_year] = comments_df[["student_id", "team_id", "nationality"]].copy()
        print(f"Nationality batch mining finished for year {target_year}.")


class XCultureVADERAnalyzer(ITextAnalyzer):
    def __init__(self):
        print(f"Initialized: VADER Analyzer for tension phrase extraction.")

    def analyze(self, comments_df: pd.DataFrame, var_week_map: dict, target_year: int, target_top_noun_phrases: int, results_container: dict):
        if comments_df.empty:
            print("comments_df is empty, skipping tension phrase extraction.")
            results_container.setdefault('yearly_vader_tension_phrases', {})[target_year] = []
            return
        phrase_data = defaultdict(lambda: {'sum_score': 0.0, 'count': 0})
        comment_columns = [col for col in var_week_map.keys() if col in comments_df.columns]
        for col in comment_columns:
            for text in comments_df[col].dropna():
                if isinstance(text, str) and text.strip():
                    blob = TextBlob(text)
                    for np_text in blob.noun_phrases:
                        phrase = np_text.lower().strip()
                        if phrase:
                            vs = vader_analyzer_instance.polarity_scores(phrase)
                            compound_score = vs['compound']
                            phrase_data[phrase]['sum_score'] += compound_score
                            phrase_data[phrase]['count'] += 1
        if not phrase_data:
            print("No noun phrases found for VADER analysis.")
            results_container.setdefault('yearly_vader_tension_phrases', {})[target_year] = []
            return
        processed_phrases = []
        for phrase, data in phrase_data.items():
            avg_score = data['sum_score'] / data['count']
            if avg_score < 0:
                processed_phrases.append((phrase, avg_score, data['count']))
        sorted_tension_phrases = sorted(processed_phrases, key=lambda x: (x[1], -x[2]))
        top_tension_phrases = sorted_tension_phrases[:target_top_noun_phrases]
        results_container.setdefault('yearly_vader_tension_phrases', {})[target_year] = top_tension_phrases


class XCultureCodebookProcessor:
    def __init__(self, codebook_file):
        self.codebook_file = codebook_file
        self.codebook_df = None
        self.var_week_map = {}
        self.weeks = []
        self._load_and_process_codebook()

    def _load_and_process_codebook(self):
        self.codebook_df = pd.read_excel(self.codebook_file, engine="openpyxl")
        self.codebook_df['Week_numeric'] = pd.to_numeric(self.codebook_df['Week'], errors='coerce')
        weekly_vars = self.codebook_df[self.codebook_df['Week_numeric'].notna()]
        weekly_vars = weekly_vars[['Variable Name', 'Week_numeric']].rename(columns={'Week_numeric': 'Week'})
        self.var_week_map = dict(zip(weekly_vars['Variable Name'], weekly_vars['Week']))
        self.weeks = sorted(set(self.var_week_map.values()))
        print("Codebook loaded and processed.")


class TeamSentimentKMeansClusteringStrategy(IClusteringStrategy):
    def __init__(self):
        print("Initialized: Team Sentiment KMeans Clustering Strategy")

    def _prepare_data_for_clustering(self, data_df, week1_sentiment_col, weekX_sentiment_col):
        if data_df.empty:
            print("data_df is empty, cannot prepare data for K-Means clustering.")
            return pd.DataFrame(), None, None

        clustering_cols = [week1_sentiment_col, weekX_sentiment_col]
        data_for_clustering = data_df[clustering_cols + ['team_id']].copy()

        initial_rows = len(data_for_clustering)
        data_for_clustering.dropna(subset=clustering_cols, inplace=True)
        rows_after_dropna = len(data_for_clustering)

        if initial_rows != rows_after_dropna:
            print(f"Dropped {initial_rows - rows_after_dropna} rows due to NaN values in sentiment columns for clustering data preparation.")

        if data_for_clustering.empty:
            print("No valid data remaining for K-Means clustering after handling NaNs.")
            return pd.DataFrame(), None, None

        features = data_for_clustering[clustering_cols]
        scaler = StandardScaler()
        scaled_features = scaler.fit_transform(features)
        print("Features scaled for K-Means clustering.")
        return data_for_clustering, scaled_features, scaler

    def perform_clustering(self, data_df: pd.DataFrame, week1_sentiment_col: str, weekX_sentiment_col: str, target_k_clusters: int, k_range_optimizer: range, var_week_map: dict = None, cohort_sentiment_data: pd.DataFrame = None) -> dict:

        results = {
            'clustered_df_target_k': pd.DataFrame(),
            'cluster_centroids_target_k': pd.DataFrame(),
            'clustered_df_optimal_k': pd.DataFrame(),
            'cluster_centroids_optimal_k': pd.DataFrame(),
            'wcss_scores': {},
            'silhouette_scores': {},
            'optimal_k_elbow': None,
            'optimal_k_silhouette': None,
            'kmeans_week1_col': week1_sentiment_col,
            'kmeans_week2_col': weekX_sentiment_col
        }

        data_for_clustering, scaled_features, scaler = self._prepare_data_for_clustering(data_df, week1_sentiment_col, weekX_sentiment_col)

        if data_for_clustering.empty or scaled_features is None or len(scaled_features) == 0:
            print("Insufficient data to perform K-Means clustering.")
            return results

        # --- Calculate WCSS and Silhouette scores for optimization ---
        wcss = {}
        silhouette_scores = {}
        for k in k_range_optimizer:
            if k == 1:  # Silhouette score is not defined for k=1
                wcss[k] = KMeans(n_clusters=k, random_state=42, n_init='auto').fit(scaled_features).inertia_
                silhouette_scores[k] = np.nan
                continue
            try:
                kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
                kmeans.fit(scaled_features)
                wcss[k] = kmeans.inertia_
                silhouette_scores[k] = silhouette_score(scaled_features, kmeans.labels_)
            except Exception as e:
                print(f"Could not calculate metrics for k={k}: {e}")
                wcss[k] = np.nan
                silhouette_scores[k] = np.nan
        print("WCSS and Silhouette scores calculated for the given k range.")
        results['wcss_scores'] = wcss
        results['silhouette_scores'] = silhouette_scores

        # --- Suggest optimal k elbow ---
        if wcss:
            # Ensure k_range_optimizer contains at least 3 points for meaningful diff_ratios, otherwise elbow method is unstable
            if len(k_range_optimizer) >= 3:
                diffs = np.diff(list(wcss.values()))
                # Use slicing to ensure diff_ratios calculation is safe for smaller diffs arrays
                # Ensure denominator is not zero
                diff_ratios = diffs[:-1] / np.where(diffs[1:] != 0, diffs[1:], np.inf) if len(diffs) > 1 else np.array([])
                diff_ratios = diff_ratios[~np.isinf(diff_ratios)]

                if len(diff_ratios) > 0:
                    optimal_k_index = np.argmax(diff_ratios) + 1
                    if optimal_k_index < len(list(wcss.keys())):
                        results['optimal_k_elbow'] = list(wcss.keys())[optimal_k_index]
                        print(f"Suggested optimal k (Elbow method): {results['optimal_k_elbow']}")
                    else:
                        print("Elbow method could not determine a stable optimal k. Consider adjusting k_range_optimizer.")
                else:
                    print("Insufficient data to determine optimal k using Elbow method or no significant bend.")
            else:
                print("Not enough k values in k_range_optimizer for Elbow method (need at least 3).")

        # --- Suggest optimal k silhouette ---
        valid_scores = {k: score for k, score in silhouette_scores.items() if not np.isnan(score)}
        if valid_scores:
            results['optimal_k_silhouette'] = max(valid_scores, key=valid_scores.get)
            print(f"Suggested optimal k (Silhouette method): {results['optimal_k_silhouette']}")
        else:
            print("No valid Silhouette scores to determine optimal k.")

        # --- Perform clustering for target_k_clusters ---
        print(f"Performing K-Means clustering with target_k_clusters={target_k_clusters}")
        kmeans_target_k = KMeans(n_clusters=target_k_clusters, random_state=42, n_init='auto')
        kmeans_target_k.fit(scaled_features)
        clustered_df_target_k = data_for_clustering.copy()
        clustered_df_target_k.loc[:, 'cluster'] = kmeans_target_k.labels_
        cluster_centers_target_k = pd.DataFrame(scaler.inverse_transform(kmeans_target_k.cluster_centers_),
                                                    columns=[week1_sentiment_col, weekX_sentiment_col])
        cluster_centers_target_k.insert(0, 'cluster', range(target_k_clusters))
        results['clustered_df_target_k'] = clustered_df_target_k
        results['cluster_centroids_target_k'] = cluster_centers_target_k

        # --- Perform clustering for optimal_k_silhouette ---
        optimal_k_silhouette_val = results['optimal_k_silhouette']
        if optimal_k_silhouette_val is not None and optimal_k_silhouette_val > 1:
            print(f"Performing K-Means clustering with optimal k (Silhouette method): {int(optimal_k_silhouette_val)}")
            kmeans_optimal_k = KMeans(n_clusters=int(optimal_k_silhouette_val), random_state=42, n_init='auto')
            kmeans_optimal_k.fit(scaled_features)
            clustered_df_optimal_k = data_for_clustering.copy()
            clustered_df_optimal_k.loc[:, 'cluster'] = kmeans_optimal_k.labels_
            cluster_centers_optimal_k = pd.DataFrame(scaler.inverse_transform(kmeans_optimal_k.cluster_centers_),
                                                     columns=[week1_sentiment_col, weekX_sentiment_col])
            cluster_centers_optimal_k.insert(0, 'cluster', range(int(optimal_k_silhouette_val)))
            results['clustered_df_optimal_k'] = clustered_df_optimal_k
            results['cluster_centroids_optimal_k'] = cluster_centers_optimal_k
        else:
            print("Optimal k from Silhouette method is not valid or not available for clustering.")

        return results


class IndividualSentimentKMeansClusteringStrategy(IClusteringStrategy):
    def __init__(self):
        print("Initialized: Individual Sentiment KMeans Clustering Strategy.")

    def _prepare_data_for_clustering(self, comments_df: pd.DataFrame, week1_sentiment_col_hint: str, weekX_sentiment_col_hint: str, var_week_map: dict, cohort_sentiment_data: pd.DataFrame):
        if comments_df.empty:
            print("comments_df is empty, cannot prepare data for K-Means clustering.")
            return pd.DataFrame(), None, None, None, None

        if 'student_id' not in comments_df.columns:
            print("Error: 'student_id' column not found in comments_df, cannot perform individual sentiment clustering.")
            return pd.DataFrame(), None, None, None, None

        if 'cohort_id' not in comments_df.columns:
            print("Error: 'cohort_id' column not found in comments_df, required for cohort sentiment merging.")
            return pd.DataFrame(), None, None, None, None

        # Extract week numbers from hints (e.g., 'week1.0_sentiment_mean' -> 1.0)
        # Note: week1_sentiment_col_hint is still passed but its individual sentiment is not used for clustering features
        # The instructions explicitly ask for (Individual Sentiment for Week X, Cohort Mean Sentiment for Week X)
        weekX_num = float(weekX_sentiment_col_hint.split('week')[1].split('.0_sentiment')[0]) if 'week' in weekX_sentiment_col_hint else 10.0
        weekX_individual_cols = [f"{var}_sentiment" for var, wk in var_week_map.items() if wk == weekX_num and f"{var}_sentiment" in comments_df.columns]

        # Dynamically find individual sentiment columns for week X based on var_week_map
        weekX_individual_cols_to_avg = [f"{var}_sentiment" for var, wk in var_week_map.items() if wk == weekX_num and f"{var}_sentiment" in comments_df.columns]

        # Filter for rows where student_id is not NaN (important before grouping)
        comments_df_valid_students = comments_df.dropna(subset=['student_id'])

        if comments_df_valid_students.empty:
            print("No valid student data after dropping NaNs in 'student_id' column.")
            return pd.DataFrame(), None, None, None, None

        # Calculate the mean individual sentiment per student_id for week X
        # Ensure we only try to average if there are columns to average
        if weekX_individual_cols_to_avg:
            individual_sentiment_weekX_series = comments_df_valid_students.groupby('student_id')[weekX_individual_cols_to_avg].mean().mean(axis=1)
        else:
            print(f"Warning: No individual sentiment columns found for Week {weekX_num}.")
            return pd.DataFrame(), None, None, None, None

        # Create a base DataFrame for clustering with student_id and this calculated individual_sentiment_weekX
        data_for_clustering = pd.DataFrame({'student_id': individual_sentiment_weekX_series.index,
                                            'individual_sentiment_weekX': individual_sentiment_weekX_series.values})

        # Merge with comments_df to get cohort_id for each student_id
        # Use a subset of comments_df for efficiency, and drop duplicates to ensure one cohort_id per student
        student_cohort_map = comments_df_valid_students[['student_id', 'cohort_id']].drop_duplicates(subset=['student_id']).set_index('student_id')
        data_for_clustering = data_for_clustering.set_index('student_id').join(student_cohort_map, how='left')
        data_for_clustering = data_for_clustering.reset_index()

        # Merge with the cohort_sentiment_data (which contains cohort_id and weekX_sentiment_mean_cohort)
        cohort_col_name = f"week{weekX_num}_sentiment_mean_cohort"
        if cohort_col_name not in cohort_sentiment_data.columns:
            print(f"Error: Cohort sentiment column '{cohort_col_name}' not found in cohort_sentiment_data.")
            return pd.DataFrame(), None, None, None, None

        # Reset index of cohort_sentiment_data for merging on 'cohort_id'
        cohort_sentiment_data_reset = cohort_sentiment_data.reset_index().rename(columns={'index': 'cohort_id'})

        data_for_clustering = data_for_clustering.merge(cohort_sentiment_data_reset[['cohort_id', cohort_col_name]], on='cohort_id', how='left')

        # Rename the column containing the cohort mean sentiment
        new_cohort_feature_name = 'cohort_mean_sentiment_weekX'
        data_for_clustering.rename(columns={cohort_col_name: new_cohort_feature_name}, inplace=True)

        clustering_cols = ['individual_sentiment_weekX', new_cohort_feature_name]

        initial_rows_pre_dropna = len(data_for_clustering)
        data_for_clustering.dropna(subset=clustering_cols, inplace=True)
        rows_after_dropna = len(data_for_clustering)

        if initial_rows_pre_dropna != rows_after_dropna:
            print(f"*** Dropped {initial_rows_pre_dropna - rows_after_dropna} rows *** due to NaN values in individual or cohort sentiment columns for clustering data preparation.")

        if data_for_clustering.empty:
            print("No valid data remaining for K-Means clustering after handling NaNs.")
            return pd.DataFrame(), None, None, None, None

        features = data_for_clustering[clustering_cols]
        scaler = StandardScaler()
        scaled_features = scaler.fit_transform(features)
        print("Features scaled for individual K-Means clustering.")
        return data_for_clustering, scaled_features, scaler, 'individual_sentiment_weekX', new_cohort_feature_name

    def perform_clustering(self, data_df: pd.DataFrame, week1_sentiment_col: str, weekX_sentiment_col: str, target_k_clusters: int, k_range_optimizer: range, var_week_map: dict = None, cohort_sentiment_data: pd.DataFrame = None) -> dict:
        # `data_df` here is expected to be `comments_df` from `XCultureData`
        # `var_week_map` and `cohort_sentiment_data` are crucial here for individual clustering.

        if var_week_map is None:
            print("Error: var_week_map is required for individual sentiment clustering but was not provided.")
            return {
                'clustered_df_target_k': pd.DataFrame(), 'cluster_centroids_target_k': pd.DataFrame(),
                'clustered_df_optimal_k': pd.DataFrame(), 'cluster_centroids_optimal_k': pd.DataFrame(),
                'wcss_scores': {}, 'silhouette_scores': {}, 'optimal_k_elbow': None, 'optimal_k_silhouette': None,
                'kmeans_week1_col': None, 'kmeans_week2_col': None
            }
        if cohort_sentiment_data is None or cohort_sentiment_data.empty:
            print("Error: cohort_sentiment_data is required for individual sentiment clustering but was not provided or is empty.")
            return {
                'clustered_df_target_k': pd.DataFrame(), 'cluster_centroids_target_k': pd.DataFrame(),
                'clustered_df_optimal_k': pd.DataFrame(), 'cluster_centroids_optimal_k': pd.DataFrame(),
                'wcss_scores': {}, 'silhouette_scores': {}, 'optimal_k_elbow': None, 'optimal_k_silhouette': None,
                'kmeans_week1_col': None, 'kmeans_week2_col': None
            }

        results = {
            'clustered_df_target_k': pd.DataFrame(),
            'cluster_centroids_target_k': pd.DataFrame(),
            'clustered_df_optimal_k': pd.DataFrame(),
            'cluster_centroids_optimal_k': pd.DataFrame(),
            'wcss_scores': {},
            'silhouette_scores': {},
            'optimal_k_elbow': None,
            'optimal_k_silhouette': None,
            'kmeans_week1_col': None,
            'kmeans_week2_col': None
        }

        prepared_data = self._prepare_data_for_clustering(
            data_df, week1_sentiment_col, weekX_sentiment_col, var_week_map, cohort_sentiment_data
        )

        # Check if _prepare_data_for_clustering returned insufficient data
        if len(prepared_data) != 5 or prepared_data[0].empty or prepared_data[1] is None or len(prepared_data[1]) == 0:
            print("Insufficient individual data to perform K-Means clustering after preparation.")
            return results

        data_for_clustering, scaled_features, scaler, individual_sentiment_weekX_col_name, cohort_mean_sentiment_weekX_col_name = prepared_data

        results['kmeans_week1_col'] = individual_sentiment_weekX_col_name # This is for plotting Y-axis
        results['kmeans_week2_col'] = cohort_mean_sentiment_weekX_col_name # This is for plotting X-axis
        clustering_cols = [individual_sentiment_weekX_col_name, cohort_mean_sentiment_weekX_col_name]

        # --- Calculate WCSS and Silhouette scores for optimization ---
        wcss = {}
        silhouette_scores = {}
        for k in k_range_optimizer:
            if k == 1:
                wcss[k] = KMeans(n_clusters=k, random_state=42, n_init='auto').fit(scaled_features).inertia_
                silhouette_scores[k] = np.nan
                continue
            try:
                kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
                kmeans.fit(scaled_features)
                wcss[k] = kmeans.inertia_
                silhouette_scores[k] = silhouette_score(scaled_features, kmeans.labels_)
            except Exception as e:
                print(f"Could not calculate metrics for k={k}: {e}")
                wcss[k] = np.nan
                silhouette_scores[k] = np.nan
        print("WCSS and Silhouette scores calculated for the given k range for individual sentiment.")
        results['wcss_scores'] = wcss
        results['silhouette_scores'] = silhouette_scores

        # --- Suggest optimal k elbow ---
        if wcss and len(k_range_optimizer) >= 3:
            diffs = np.diff(list(wcss.values()))
            # Ensure denominator is not zero before division
            diff_ratios = diffs[:-1] / np.where(diffs[1:] != 0, diffs[1:], np.inf) if len(diffs) > 1 else np.array([])
            diff_ratios = diff_ratios[~np.isinf(diff_ratios)] # Remove inf values

            if len(diff_ratios) > 0:
                optimal_k_index = np.argmax(diff_ratios) + 1
                if optimal_k_index < len(list(wcss.keys())):
                    results['optimal_k_elbow'] = list(wcss.keys())[optimal_k_index]
                    print(f"Suggested optimal k (Elbow method) for individual sentiment: {results['optimal_k_elbow']}")
                else:
                    print("Elbow method could not determine a stable optimal k for individual sentiment. Consider adjusting k_range_optimizer.")
            else:
                print("Insufficient data to determine optimal k using Elbow method or no significant bend for individual sentiment.")
        else:
            print("Not enough k values in k_range_optimizer for Elbow method for individual sentiment.")

        # --- Suggest optimal k silhouette ---
        valid_scores = {k: score for k, score in silhouette_scores.items() if not np.isnan(score)}
        if valid_scores:
            results['optimal_k_silhouette'] = max(valid_scores, key=valid_scores.get)
            print(f"Suggested optimal k (Silhouette method) for individual sentiment: {results['optimal_k_silhouette']}")
        else:
            print("No valid Silhouette scores to determine optimal k for individual sentiment.")

        # --- Perform clustering for target_k_clusters ---
        print(f"Performing K-Means clustering with target_k_clusters={target_k_clusters} for individual sentiment.")
        kmeans_target_k = KMeans(n_clusters=target_k_clusters, random_state=42, n_init='auto')
        kmeans_target_k.fit(scaled_features)
        clustered_df_target_k = data_for_clustering.copy()
        clustered_df_target_k.loc[:, 'cluster'] = kmeans_target_k.labels_
        cluster_centers_target_k = pd.DataFrame(scaler.inverse_transform(kmeans_target_k.cluster_centers_),
                                                    columns=clustering_cols)
        cluster_centers_target_k.insert(0, 'cluster', range(target_k_clusters))
        results['clustered_df_target_k'] = clustered_df_target_k
        results['cluster_centroids_target_k'] = cluster_centers_target_k

        # --- Perform clustering for optimal_k_silhouette ---
        optimal_k_silhouette_val = results['optimal_k_silhouette']
        if optimal_k_silhouette_val is not None and optimal_k_silhouette_val > 1:
            print(f"Performing K-Means clustering with optimal k (Silhouette method): {int(optimal_k_silhouette_val)} for individual sentiment.")
            kmeans_optimal_k = KMeans(n_clusters=int(optimal_k_silhouette_val), random_state=42, n_init='auto')
            kmeans_optimal_k.fit(scaled_features)
            clustered_df_optimal_k = data_for_clustering.copy()
            clustered_df_optimal_k.loc[:, 'cluster'] = kmeans_optimal_k.labels_
            cluster_centers_optimal_k = pd.DataFrame(scaler.inverse_transform(kmeans_optimal_k.cluster_centers_),
                                                     columns=clustering_cols)
            cluster_centers_optimal_k.insert(0, 'cluster', range(int(optimal_k_silhouette_val)))
            results['clustered_df_optimal_k'] = clustered_df_optimal_k
            results['cluster_centroids_optimal_k'] = cluster_centers_optimal_k
        else:
            print("Optimal k from Silhouette method is not valid or not available for individual sentiment clustering.")

        return results


class XCultureData:
    def __init__(self, comments_df_base, codebook_df, var_week_map, weeks, text_analyzers: list[ITextAnalyzer], target_year=None, target_top_noun_phrases=30):
        self.comments_df = comments_df_base.copy()
        self.codebook_df = codebook_df
        self.var_week_map = var_week_map
        self.weeks = weeks
        self.text_analyzers = text_analyzers
        self.target_year = target_year
        self.timenow = datetime.now()
        self.sheet_name_with_date = self.timenow.strftime("%Y%m%d%H%M")
        self.yearly_top_noun_phrases = {}
        self.yearly_yake_keywords = {}
        self.yearly_vader_tension_phrases = {}
        self.yearly_nationality_results = {}
        self.yearly_cohort_sentiment = {}
        self.target_top_noun_phrases = target_top_noun_phrases
        self._apply_year_filter()
        if 'team_id' in self.comments_df.columns:
            self.comments_df['cohort_id'] = self.comments_df['team_id'].astype(str).apply(
                lambda x: f'{x.split("-")[1]}-{x.split("-")[2]}' if pd.notna(x) and x.count('-') >= 2 else pd.NA
            )
            print("Added 'cohort_id' column based on 'team_id' (format: Part2-Part3).")
        else:
            print("Warning: 'team_id' column not found. 'cohort_id' not created.")

        self._sanitise_comments()
        results_container = {
            'yearly_top_noun_phrases': self.yearly_top_noun_phrases,
            'yearly_yake_keywords': self.yearly_yake_keywords,
            'yearly_vader_tension_phrases': self.yearly_vader_tension_phrases,
            'yearly_nationality_results': self.yearly_nationality_results
        }
        for analyzer in self.text_analyzers:
            analyzer.analyze(self.comments_df, self.var_week_map, self.target_year, self.target_top_noun_phrases, results_container)
        self._compute_individual_weekly_scores()
        self._compute_cohort_weekly_sentiment()

    def _apply_year_filter(self):
        if self.target_year is not None and 'analysis_year' in self.comments_df.columns:
            self.comments_df = self.comments_df[self.comments_df['analysis_year'] == self.target_year].copy()

    @staticmethod
    def _sanitise_text(text):
        if pd.isna(text) or not isinstance(text, str) or text.strip() == "": return ""
        cleaned_text = text.lower()
        keywords_to_remove = ['x-culture\'s','x culture\'s','xculture\'s','x-culture','x culture','xculture']
        for kw in keywords_to_remove: cleaned_text = cleaned_text.replace(kw, '')
        return re.sub(r'[^\x20-\x7E]', '', cleaned_text)

    def _sanitise_comments(self):
        comment_columns = [col for col in self.var_week_map.keys() if col in self.comments_df.columns]
        for col in comment_columns:
            self.comments_df[col] = self.comments_df[col].apply(self._sanitise_text)

    @staticmethod
    def _analyse_text(text):
        if not text: return pd.Series({'sentiment': None, 'subjectivity': None, 'vader': None, 'vader_neg': None, 'vader_neu': None, 'vader_pos': None, 'word_count': 0})
        blob = TextBlob(text)
        stop_words = set(stopwords.words('english'))
        filtered_words = [word for word in blob.words if word.isalpha() and word not in stop_words]
        vader_score = vader_analyzer_instance.polarity_scores(text)['compound']
        vader_score_negative = vader_analyzer_instance.polarity_scores(text)['neg']
        vader_score_neutral = vader_analyzer_instance.polarity_scores(text)['neu']
        vader_score_positive = vader_analyzer_instance.polarity_scores(text)['pos']
        return pd.Series({
            'sentiment': blob.sentiment.polarity,
            'subjectivity': blob.sentiment.subjectivity,
            'vader': vader_score,
            'vader_neg': vader_score_negative,
            'vader_neu': vader_score_neutral,
            'vader_pos': vader_score_positive,
            'word_count': len(filtered_words)
        })

    def _compute_individual_weekly_scores(self):
        if self.comments_df.empty: return
        new_cols = []
        print(f"Processing Text Analysis and VADER Scoring...")
        for col in self.var_week_map.keys():
            if col in self.comments_df.columns:
                res = self.comments_df[col].apply(XCultureData._analyse_text)
                new_cols.append(res['sentiment'].rename(f"{col}_sentiment"))
                new_cols.append(res['subjectivity'].rename(f"{col}_subjectivity"))
                new_cols.append(res['vader'].rename(f"{col}_vader"))
                new_cols.append(res['vader_neg'].rename(f"{col}_vader_neg"))
                new_cols.append(res['vader_neu'].rename(f"{col}_vader_neu"))
                new_cols.append(res['vader_pos'].rename(f"{col}_vader_pos"))
                new_cols.append(res['word_count'].rename(f"{col}_word_count"))
        if new_cols: self.comments_df = pd.concat([self.comments_df] + new_cols, axis=1)

        weekly_series = []
        for week in self.weeks:
            for metric in ['sentiment', 'subjectivity', 'vader', 'word_count']:
                metric_cols = [f"{v}_{metric}" for v, wk in self.var_week_map.items() if wk == week and f"{v}_{metric}" in self.comments_df.columns]
                if metric_cols:
                    if metric == 'word_count':
                        weekly_series.append(self.comments_df[metric_cols].sum(axis=1).rename(f"week{week}_{metric}"))
                    else:
                        weekly_series.append(self.comments_df[metric_cols].mean(axis=1).rename(f"week{week}_{metric}"))
                else:
                    weekly_series.append(pd.Series(pd.NA, index=self.comments_df.index, name=f"week{week}_{metric}"))
        if weekly_series: self.comments_df = pd.concat([self.comments_df] + weekly_series, axis=1)

    def _compute_cohort_weekly_sentiment(self):
        if self.comments_df.empty or 'cohort_id' not in self.comments_df.columns:
            print("comments_df is empty or 'cohort_id' is missing, skipping cohort weekly sentiment computation.")
            self.yearly_cohort_sentiment[self.target_year] = pd.DataFrame() # Store an empty DataFrame
            return

        cohort_weekly_sentiment_data = []

        for week in self.weeks:
            individual_sentiment_col = f"week{week}_sentiment"

            if individual_sentiment_col in self.comments_df.columns:
                cohort_mean_sentiment = self.comments_df.groupby('cohort_id')[individual_sentiment_col].mean().rename(f"week{week}_sentiment_mean_cohort")
                cohort_weekly_sentiment_data.append(cohort_mean_sentiment)
            else:
                print(f"Warning: Individual sentiment column '{individual_sentiment_col}' not found for cohort sentiment calculation.")

        if cohort_weekly_sentiment_data:
            cohort_sentiment_df = pd.concat(cohort_weekly_sentiment_data, axis=1)
            self.yearly_cohort_sentiment[self.target_year] = cohort_sentiment_df
            print(f"Cohort weekly sentiment computed for year {self.target_year}.")
        else:
            self.yearly_cohort_sentiment[self.target_year] = pd.DataFrame() # Store an empty DataFrame if no data
            print(f"No cohort weekly sentiment data found for year {self.target_year}.")


class XCultureTeamAnalysis:
    def __init__(self, comments_df, var_week_map, weeks):
        self.comments_df = comments_df
        self.var_week_map = var_week_map
        self.weeks = weeks
        self.team_summary = pd.DataFrame()
        self.top_10_teams_summary = pd.DataFrame()
        self.lowest_10_teams_summary = pd.DataFrame()

        self._aggregate_team_scores()
        if not self.team_summary.empty:
            self._identify_top_teams()
            self._identify_lowest_teams()
        else:
            print("Team summary is empty, skipping identification of top/lowest teams.")

    def _aggregate_team_scores(self):
        potential_individual_weekly_score_cols = []
        for wk in self.weeks:
            potential_individual_weekly_score_cols.append(f"week{wk}_sentiment")
            potential_individual_weekly_score_cols.append(f"week{wk}_subjectivity")
            # potential_individual_weekly_score_cols.append(f"week{wk}_WC")
            potential_individual_weekly_score_cols.append(f"week{wk}_word_count")

        existing_individual_weekly_score_cols = [col for col in potential_individual_weekly_score_cols if col in self.comments_df.columns]

        agg_dict = {}
        for col in existing_individual_weekly_score_cols:
            agg_dict[f"{col}_mean"] = (col, 'mean')
            agg_dict[f"{col}_std"] = (col, 'std')

        if self.comments_df.empty:
            print("comments_df is empty, skipping team score aggregation.")
            empty_cols = ['team_id'] + list(agg_dict.keys())
            self.team_summary = pd.DataFrame(columns=empty_cols)
            return

        if 'team_id' in self.comments_df.columns and agg_dict:
            team_member_counts = self.comments_df.groupby('team_id').size().rename('team_member_count')
            self.team_summary = self.comments_df.groupby('team_id').agg(**agg_dict).reset_index()
            self.team_summary = self.team_summary.merge(team_member_counts, on='team_id', how='left')
            cols_to_check = [col for col in self.team_summary.columns if col.endswith('_mean') or col.endswith('_std')]
            for col in cols_to_check:
                self.team_summary.loc[self.team_summary['team_member_count'] < 2, col] = pd.NA
            # self.team_summary.drop(columns=['team_member_count'], inplace=True)

        else:
            print("Warning: No valid columns to aggregate or 'team_id' not found. team_summary will be empty.")
            empty_cols = ['team_id'] + list(agg_dict.keys())
            self.team_summary = pd.DataFrame(columns=empty_cols)

        if 'nationality' in self.comments_df.columns and agg_dict:
            team_member_nationality_list = self.comments_df.groupby('team_id')['nationality'].agg(list).reset_index()
            self.team_summary = self.team_summary.merge(team_member_nationality_list, on='team_id', how='outer')
            # print(f"DEBUG: Sample team summary entries: {list(self.team_summary.items())[:5]}")

        else:
            print("Warning: No valid columns to aggregate or 'nationality2014' not found. team_summary will be empty.")
            empty_cols = ['team_id'] + list(agg_dict.keys())
            self.team_summary = pd.DataFrame(columns=empty_cols)


    def _identify_top_teams(self):
        sentiment_col_week1 = 'week1.0_sentiment_mean'
        if sentiment_col_week1 in self.team_summary.columns and not self.team_summary[sentiment_col_week1].isnull().all():
            self.top_10_teams_summary = self.team_summary.dropna(subset=[sentiment_col_week1])\
                                             .sort_values(by=sentiment_col_week1, ascending=False)\
                                             .head(10)
        else:
            print(f"Warning: '{sentiment_col_week1}' not found or all values are NaN. Cannot identify top teams.")

    def _identify_lowest_teams(self):
        sentiment_col_week1 = 'week1.0_sentiment_mean'
        if sentiment_col_week1 in self.team_summary.columns and not self.team_summary[sentiment_col_week1].isnull().all():
            self.lowest_10_teams_summary = self.team_summary.dropna(subset=[sentiment_col_week1])\
                                              .sort_values(by=sentiment_col_week1, ascending=True)\
                                              .head(10)
        else:
            print(f"Warning: '{sentiment_col_week1}' not found or all values are NaN. Cannot identify lowest teams.")


class XCultureExporter:
    def __init__(self, target_year=None):
        self.timenow = datetime.now()
        self.sheet_name_with_date = self.timenow.strftime("%Y%m%d%H%M")
        self.target_year = target_year

    def export_dataframe_to_excel(self, dataframe, output_filename, sheet_name_prefix="Sheet1"):
        if dataframe.empty:
            print(f"Warning: DataFrame is empty. Skipping export of '{sheet_name_prefix}' to {output_filename}.")
            return
        if self.target_year:
            sheet_name = f"{sheet_name_prefix}_{self.target_year}"
        else:
            sheet_name = f"{sheet_name_prefix}_{self.sheet_name_with_date}"
        try:
            file_exists = os.path.exists(output_filename)
            mode = 'a' if file_exists else 'w'

            writer_kwargs = {'engine': 'openpyxl', 'mode': mode}
            if mode == 'a':
                writer_kwargs['if_sheet_exists'] = 'replace'

            with pd.ExcelWriter(output_filename, **writer_kwargs) as writer:
                dataframe.to_excel(writer, sheet_name=sheet_name, index=False)
            print(f"\u2705 Data successfully exported to {output_filename} on sheet '{sheet_name}'.")
        except Exception as e:
            print(f"Error exporting data to {output_filename}: {e}")

    def export_multiple_dataframes_to_excel(self, dataframes_dict, output_filename):
        if not dataframes_dict:
            print(f"Warning: No dataframes provided for export to {output_filename}.")
            return

        file_exists = os.path.exists(output_filename)
        mode = 'a' if file_exists else 'w'

        writer_kwargs = {'engine': 'openpyxl', 'mode': mode}
        if mode == 'a':
                writer_kwargs['if_sheet_exists'] = 'replace'

        with pd.ExcelWriter(output_filename, **writer_kwargs) as writer:
            for sheet_name_prefix, dataframe in dataframes_dict.items():
                if dataframe.empty:
                    print(f"    - Warning: DataFrame for sheet '{sheet_name_prefix}' is empty. Skipping export.")
                    continue
                if self.target_year:
                    sheet_name = f"{sheet_name_prefix}_{self.target_year}"
                else:
                    sheet_name = f"{sheet_name_prefix}_{self.sheet_name_with_date}"
                try:
                    dataframe.to_excel(writer, sheet_name=sheet_name, index=False)
                except Exception as e:
                    print(f"    - Error exporting sheet '{sheet_name}' to {output_filename}: {e}")
        print(f"\u2705 All specified dataframes successfully exported to {output_filename}.")


# --- IPlottingStrategy Implementations ---

class SentimentAndSubjectivityPlotter(IPlottingStrategy):
    def __init__(self):
        print(f"Initialized: Sentiment And Subjectivity Plotter.")

    def plot(self, pdf_pages, data: dict):
        dataframe = data['dataframe']
        title = data['title']
        plot_type = data['plot_type']
        weeks_to_plot = data['weeks_to_plot']
        cohort_sentiment = data['cohort_sentiment']
        target_year = data['target_year']

        if dataframe.empty or not weeks_to_plot:
            print(f"No data to visualize for {title} {plot_type}.")
            return

        fig = plt.figure(figsize=(12, 6))
        for _, row in dataframe.iterrows():
            data_to_plot = []
            weeks_for_plot = []
            for wk in weeks_to_plot:
                col_name_mean = f"week{wk}_{plot_type}_mean"
                if col_name_mean in row.index and pd.notna(row[col_name_mean]):
                    data_to_plot.append(row[col_name_mean])
                    weeks_for_plot.append(wk)
            if data_to_plot:
                plt.plot(weeks_for_plot, data_to_plot, marker='o', alpha=0.6, label=row['team_id'])

        if not cohort_sentiment.empty:
            cohort_ids = sorted(cohort_sentiment.index.unique())
            num_cohorts = len(cohort_ids)
            # FIX: Use matplotlib.colormaps.get_cmap directly
            cohort_colors = plt.colormaps.get_cmap('Dark2') # Corrected line
            cohort_linestyles = ['--', '-.', ':', (0, (3, 5, 1, 5, 1, 5)), (0, (3, 1, 1, 1))]

            for i, cohort_id in enumerate(cohort_ids):
                row = cohort_sentiment.loc[cohort_id]
                cohort_data = []
                weeks_for_cohort_plot = []
                for wk in weeks_to_plot:
                    col_name = f"week{wk}_{plot_type}_mean_cohort"
                    if col_name in row.index and pd.notna(row[col_name]):
                        cohort_data.append(row[col_name])
                        weeks_for_cohort_plot.append(wk)
                if cohort_data:
                    plt.plot(weeks_for_cohort_plot, cohort_data,
                             linestyle=cohort_linestyles[i % len(cohort_linestyles)],
                             color=cohort_colors(i),
                             alpha=0.8,
                             label=f'Cohort {cohort_id} {plot_type.capitalize()}')

        plt.title(title)
        plt.xlabel("Week")
        plt.ylabel(f"Average {plot_type.capitalize()} (Mean)")
        if plot_type == 'sentiment':
            plt.axhline(0, color='grey', linestyle='--')
            plt.ylim(-1, 1)
        else:
            plt.axhline(0.5, color='grey', linestyle='--')
            plt.ylim(0, 1)
        plt.grid(True)
        plt.xticks([int(w) for w in weeks_to_plot])
        plt.legend(title='ID', bbox_to_anchor=(1.0, 1), loc='upper left')
        plt.tight_layout(pad=0.5)
        pdf_pages.savefig(fig)
        plt.close(fig)

class SentimentWordCountPlotter(IPlottingStrategy):
    def __init__(self):
        print(f"Initialized: Sentiment Word Count Plotter.")

    def plot(self, pdf_pages, data: dict):
        dataframe = data['dataframe']
        title = data['title']
        weeks_to_plot = data['weeks_to_plot']
        global_max_word_count = data['global_max_word_count']
        cohort_sentiment = data['cohort_sentiment']
        target_year = data['target_year'] # Added for consistent data structure

        if dataframe.empty or not weeks_to_plot:
            print(f"No data to visualize for {title} (Sentiment & Word Count).")
            return

        fig = plt.figure(figsize=(14, 7))
        ax1 = plt.gca()
        ax2 = ax1.twinx()

        ax1.set_xlabel("Week")
        ax1.set_ylabel("Average Sentiment (Mean)", color='blue')
        ax2.set_ylabel("Average Word Count (Mean)", color='black')

        ax1.tick_params(axis='y', labelcolor='blue')
        ax2.tick_params(axis='y', labelcolor='black')

        all_lines = []
        all_labels = []

        # FIX: Use matplotlib.colormaps.get_cmap directly
        colors = plt.colormaps.get_cmap('tab10') # Corrected line
        markers = ['o', 's', '^', 'D', 'v', 'p', '*', 'h', 'P', 'X']

        for i, (_, row) in enumerate(dataframe.iterrows()):
            sentiment_data = []
            word_count_data = []
            weeks_for_plot = []
            for wk in weeks_to_plot:
                sentiment_col = f"week{wk}_sentiment_mean"
                word_count_col = f"week{wk}_word_count_mean"
                if sentiment_col in row.index and pd.notna(row[sentiment_col]) and pd.notna(row[word_count_col]):
                    sentiment_data.append(row[sentiment_col])
                    word_count_data.append(row[word_count_col])
                    weeks_for_plot.append(wk)

            if weeks_for_plot:
                team_color = colors(i)
                team_marker = markers[i % len(markers)]

                line1, = ax1.plot(weeks_for_plot, sentiment_data, marker=team_marker, linestyle='-', alpha=0.8, color=team_color, label=f"{row['team_id']} Sentiment")
                line2, = ax2.plot(weeks_for_plot, word_count_data, marker=team_marker, linestyle='--', alpha=0.5, color=team_color, label=f"{row['team_id']} Word Count")

                all_lines.extend([line1, line2])
                all_labels.extend([f"{row['team_id']} Sentiment", f"{row['team_id']} Word Count"])

        if not cohort_sentiment.empty:
            cohort_ids = sorted(cohort_sentiment.index.unique())
            num_cohorts = len(cohort_ids)
            # FIX: Use matplotlib.colormaps.get_cmap directly
            cohort_colors = plt.colormaps.get_cmap('Dark2') # Corrected line
            cohort_linestyles = ['--', '-.', ':', (0, (3, 5, 1, 5, 1, 5)), (0, (3, 1, 1, 1))]

            for i, cohort_id in enumerate(cohort_ids):
                row = cohort_sentiment.loc[cohort_id]
                cohort_sentiment_data = []
                weeks_for_cohort_plot = [] # Initialize here
                # Hardcoding to 'sentiment' as this plotter shows sentiment vs word count.
                plot_type_for_cohort = 'sentiment'
                for wk in weeks_to_plot:
                    col_name = f"week{wk}_{plot_type_for_cohort}_mean_cohort"
                    if col_name in row.index and pd.notna(row[col_name]):
                        cohort_sentiment_data.append(row[col_name])
                        weeks_for_cohort_plot.append(wk)
                if cohort_sentiment_data:
                    cohort_line, = ax1.plot(weeks_for_cohort_plot, cohort_sentiment_data,
                                            linestyle=cohort_linestyles[i % len(cohort_linestyles)],
                                            color=cohort_colors(i),
                                            alpha=0.8,
                                            label=f'Cohort {cohort_id} Sentiment')
                    all_lines.append(cohort_line)
                    all_labels.append(f'Cohort {cohort_id} Sentiment')

        ax1.axhline(0, color='grey', linestyle='--', linewidth=0.7)
        ax1.set_ylim(-1, 1)
        ax2.set_ylim(0, global_max_word_count)

        plt.title(title)
        ax1.grid(True, linestyle=':', alpha=0.6)

        plt.xticks([int(w) for w in weeks_to_plot])

        ax2.legend(all_lines, all_labels, title='ID & Metric', bbox_to_anchor=(1.0, 1), loc='upper left')

        plt.tight_layout(pad=0.5)
        pdf_pages.savefig(fig)
        plt.close(fig)

class NounPhraseReporter(IPlottingStrategy):
    def __init__(self):
        print(f"Initialized: Noun Phrase Reporter.")

    def plot(self, pdf_pages, data: dict):
        noun_phrases = data['noun_phrases']
        target_year = data['target_year']
        target_top_noun_phrases = data['target_top_noun_phrases']

        if target_year in noun_phrases and noun_phrases[target_year]:
            fig, ax = plt.subplots(figsize=(8.27, 11.69))
            ax.axis('off')
            ax.set_title(f'Top {target_top_noun_phrases} Noun Phrases for Year {target_year}', fontsize=10, pad=10)

            noun_phrases_text = []
            for i, (phrase, count) in enumerate(noun_phrases[target_year]):
                noun_phrases_text.append(f"{i+1}. {phrase.capitalize()}: {count}")

            if len(noun_phrases_text) > 25:
                mid_point = len(noun_phrases_text) // 2
                col1 = noun_phrases_text[:mid_point]
                col2 = noun_phrases_text[mid_point:]

                ax.text(0.1, 0.9, "\n".join(col1), ha='left', va='top', transform=ax.transAxes, fontsize=10)
                ax.text(0.5, 0.9, "\n".join(col2), ha='left', va='top', transform=ax.transAxes, fontsize=10)
            else:
                ax.text(0.1, 0.9, "\n".join(noun_phrases_text), ha='left', va='top', transform=ax.transAxes, fontsize=10)

            pdf_pages.savefig(fig)
            plt.close(fig)
        else:
            print(f"No noun phrases available for year {target_year} to write to PDF.")

class YAKEKeywordReporter(IPlottingStrategy):
    def __init__(self):
        print(f"Initialized: YAKE Keyword Reporter.")

    def plot(self, pdf_pages, data: dict):
        yake_keywords = data['yake_keywords']
        target_year = data['target_year']
        target_top_noun_phrases = data['target_top_noun_phrases']

        if target_year in yake_keywords and yake_keywords[target_year]:
            fig, ax = plt.subplots(figsize=(8.27, 11.69))
            ax.axis('off')
            ax.set_title(f'Top {target_top_noun_phrases} YAKE Keywords for Year {target_year}', fontsize=10, pad=10)

            yake_keywords_text = []
            for i, (keyword, score) in enumerate(yake_keywords[target_year]):
                yake_keywords_text.append(f"{i+1}. {keyword.capitalize()}: {score:.4f}")

            if len(yake_keywords_text) > 25:
                mid_point = len(yake_keywords_text) // 2
                col1 = yake_keywords_text[:mid_point]
                col2 = yake_keywords_text[mid_point:]

                ax.text(0.0, 0.9, "\n".join(col1), ha='left', va='top', transform=ax.transAxes, fontsize=9, bbox=dict(facecolor='red', alpha=0.2))
                ax.text(0.5, 0.9, "\n".join(col2), ha='left', va='top', transform=ax.transAxes, fontsize=9, bbox=dict(facecolor='yellow', alpha=0.2))
            else:
                ax.text(0.1, 0.9, "\n".join(yake_keywords_text), ha='left', va='top', transform=ax.transAxes, fontsize=10)

            pdf_pages.savefig(fig)
            plt.close(fig)
        else:
            print(f"No YAKE keywords available for year {target_year} to write to PDF.")

class VADERCompoundPhraseReporter(IPlottingStrategy):
    def __init__(self):
        print(f"Initialized: VADER Compound Phrase Reporter.")

    def plot(self, pdf_pages, data: dict):
        vader_tension_phrases = data['vader_tension_phrases']
        target_year = data['target_year']
        target_top_noun_phrases = data['target_top_noun_phrases']

        if target_year in vader_tension_phrases and vader_tension_phrases[target_year]:
            fig, ax = plt.subplots(figsize=(8.27, 11.69)) # Correctly unpack fig, ax
            ax.axis('off')
            ax.set_title(f'Top {target_top_noun_phrases} VADER Tension Phrases for Year {target_year}', fontsize=10, pad=10)

            vader_phrases_text = []
            for i, (phrase, score, frequency) in enumerate(vader_tension_phrases[target_year]):
                vader_phrases_text.append(f"{i+1}. {phrase.capitalize()}: S {score:.4f}, F {frequency}")

            if len(vader_phrases_text) > 25:
                mid_point = len(vader_phrases_text) // 2
                col1 = vader_phrases_text[:mid_point]
                col2 = vader_phrases_text[mid_point:]

                ax.text(0.0, 0.9, "\n".join(col1), ha='left', va='top', transform=ax.transAxes, fontsize=9)
                ax.text(0.5, 0.9, "\n".join(col2), ha='left', va='top', transform=ax.transAxes, fontsize=9)
            else:
                ax.text(0.1, 0.9, "\n".join(vader_phrases_text), ha='left', va='top', transform=ax.transAxes, fontsize=10)

            pdf_pages.savefig(fig)
            plt.close(fig)
        else:
            print(f"No VADER tension phrases available for year {target_year} to write to PDF.")

class KMeansClusterPlotter(IPlottingStrategy):
    def __init__(self):
        print(f"initialized: KMeansClusterPlotter.")

    def plot(self, pdf_pages, data: dict):
        clustered_df = data['clustered_df']
        centroids_df = data['centroids_df']
        kmeans_week1_col = data['kmeans_week1_col']
        kmeans_week2_col = data['kmeans_week2_col']
        title_suffix = data['title_suffix']
        target_year = data['target_year']
        id_col = data.get('id_col', 'team_id') # Default to team_id, can be student_id
        id_col_type = data.get('id_col_type', 'Team') # Get id_col_type for consistent title

        if clustered_df.empty or centroids_df.empty or kmeans_week1_col is None or kmeans_week2_col is None:
            print(f"No K-Means cluster data or column information to visualize {title_suffix} for year {target_year}.")
            return

        fig, ax = plt.subplots(figsize=(11.69, 8.27))

        # Dynamically determine axis labels based on the clustering type
        if id_col_type == 'Individual':
            # For individual clustering with cohort comparison
            # Extract week number from kmeans_week2_col for labeling
            week_num_str = re.search(r'week(\d+\.0)', kmeans_week2_col)
            week_label = f" (Week {week_num_str.group(1).replace('.0', '')})" if week_num_str else " (Latest Week)"
            week1_label = f"Individual Sentiment{week_label}"
            week2_label = f"Cohort Mean Sentiment{week_label}"
        else:
            # For team clustering or other generic cases
            week1_label = kmeans_week1_col.replace("_individual_sentiment", " Individual Sentiment").replace("_sentiment_mean", " Team Sentiment").replace("week", "Week ").replace(".0", "")
            week2_label = kmeans_week2_col.replace("_individual_sentiment", " Individual Sentiment").replace("_sentiment_mean", " Team Sentiment").replace("week", "Week ").replace(".0", "")

        for cluster_id in sorted(clustered_df['cluster'].unique()):
            cluster_data = clustered_df[clustered_df['cluster'] == cluster_id]
            ax.scatter(
                cluster_data[kmeans_week2_col], # X-axis
                cluster_data[kmeans_week1_col],  # Y-axis
                label=f'Cluster {int(cluster_id)} (N={len(cluster_data)})',
                alpha=0.7
            )

        if not centroids_df.empty:
            ax.scatter(
                centroids_df[kmeans_week2_col],
                centroids_df[kmeans_week1_col],
                marker='X',
                s=200,
                color='red',
                edgecolors='white',
                label='Cluster Centroids'
            )
            for idx, row in centroids_df.iterrows():
                ax.annotate(
                    f'C{int(row["cluster"])}',
                    (row[kmeans_week2_col], row[kmeans_week1_col]),
                    textcoords="offset points",
                    xytext=(0,10),
                    ha='center',
                    color='red',
                    fontsize=10,
                    fontweight='bold'
                )

        all_sentiment_values = pd.concat([
            clustered_df[kmeans_week1_col],
            clustered_df[kmeans_week2_col],
            centroids_df[kmeans_week1_col],
            centroids_df[kmeans_week2_col]
        ])
        min_val = all_sentiment_values.min() - 0.1
        max_val = all_sentiment_values.max() + 0.1

        ax.set_xlim(min_val, max_val)
        ax.set_ylim(min_val, max_val)

        ax.set_title(f'K-Means Clustering of {id_col_type} Sentiment (Year {target_year}) {title_suffix}') # Use id_col_type for consistent title
        ax.set_xlabel(week2_label) # Use dynamic label
        ax.set_ylabel(week1_label)  # Use dynamic label
        ax.grid(True, linestyle=':', alpha=0.6)

        # Add the diagonal reference line for individual clustering
        if id_col_type == 'Individual':
            # Plot a diagonal line from the minimum to the maximum of the combined sentiment range
            diag_min = min(ax.get_xlim()[0], ax.get_ylim()[0])
            diag_max = max(ax.get_xlim()[1], ax.get_ylim()[1])
            ax.plot([diag_min, diag_max], [diag_min, diag_max], ls="--", c="darkgrey", label="Individual = Cohort Mean", zorder=0) # Add zorder to ensure it's behind points

        ax.legend(title='Clusters', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
        plt.tight_layout(rect=[0, 0, 0.85, 1])

        pdf_pages.savefig(fig)
        plt.close(fig)
        print(f"\u2705 K-Means clustering plot {title_suffix} for year {target_year} saved to PDF.")

class WCSSSilhouettePlotter(IPlottingStrategy):
    def __init__(self):
        print(f"Initialized: WCSS Silhouette Plotter.")

    def plot(self, pdf_pages, data: dict):
        wcss_scores = data['wcss_scores']
        silhouette_scores = data['silhouette_scores']
        optimal_k_elbow = data['optimal_k_elbow']
        optimal_k_silhouette = data['optimal_k_silhouette']
        target_year = data['target_year']
        id_col_type = data.get('id_col_type', 'Team') # 'Team' or 'Individual'

        if not wcss_scores or not silhouette_scores:
            print("No WCSS or Silhouette scores to plot.")
            return

        k_values_wcss = sorted(wcss_scores.keys())
        wcss_values = [wcss_scores[k] for k in k_values_wcss]

        k_values_silhouette = []
        silhouette_values = []
        for k in sorted(silhouette_scores.keys()):
            score = silhouette_scores[k]
            if not np.isnan(score):
                k_values_silhouette.append(k)
                silhouette_values.append(score)

        fig, ax1 = plt.subplots(figsize=(11.69, 8.27))

        line1, = ax1.plot(k_values_wcss, wcss_values, marker='o', color='blue', label='WCSS')
        ax1.set_xlabel('Number of Clusters (k)')
        ax1.set_ylabel('Within-Cluster Sum of Squares (WCSS)', color='blue')
        ax1.tick_params(axis='y', labelcolor='blue')
        ax1.grid(True, linestyle=':', alpha=0.6)

        ax2 = ax1.twinx()
        line2, = ax2.plot(k_values_silhouette, silhouette_values, marker='x', color='green', label='Silhouette Score')
        ax2.set_ylabel('Silhouette Score', color='black')
        ax2.tick_params(axis='y', labelcolor='black')

        all_lines_legend = [line1, line2]
        all_labels_legend = ['WCSS', 'Silhouette Score']

        if optimal_k_elbow is not None and optimal_k_elbow in wcss_scores:
            elbow_scatter = ax1.scatter(
                optimal_k_elbow,
                wcss_scores[optimal_k_elbow],
                marker='*', s=200,
                color='black',
                edgecolors='red',
                label=f'Elbow k={int(optimal_k_elbow)} (WCSS)'
            )
            all_lines_legend.append(elbow_scatter)
            all_labels_legend.append(f'Elbow k={int(optimal_k_elbow)} (WCSS)')

        if optimal_k_silhouette is not None and optimal_k_silhouette in silhouette_scores and not np.isnan(silhouette_scores[optimal_k_silhouette]):
            silhouette_scatter = ax2.scatter(
                optimal_k_silhouette,
                silhouette_scores[optimal_k_silhouette],
                marker='D', s=100,
                color='black',
                edgecolors='red',
                label=f'Silhouette k={int(optimal_k_silhouette)} (Score)'
            )
            all_lines_legend.append(silhouette_scatter)
            all_labels_legend.append(f'Silhouette k={int(optimal_k_silhouette)} (Score)')

        plt.title(f'{id_col_type} WCSS and Silhouette Score for Optimal k (Year {target_year})')

        ax2.legend(all_lines_legend, all_labels_legend, loc='upper right')

        plt.tight_layout()
        pdf_pages.savefig(fig)
        plt.close(fig)
        print(f"\u2705 Combined {id_col_type} WCSS and Silhouette plot for year {target_year} saved to PDF.")

class CohortSentimentPlotter(IPlottingStrategy):
    def __init__(self):
        print(f"Initialized: Cohort Sentiment Plotter.")

    def plot(self, pdf_pages, data: dict):
        cohort_sentiment = data['cohort_sentiment']
        target_year = data['target_year']
        weeks_to_plot = data['weeks_to_plot']

        if cohort_sentiment.empty:
            print(f"No cohort sentiment data to plot for year {target_year}.")
            return

        fig = plt.figure(figsize=(12, 6))
        cohort_sentiment_cols = [col for col in cohort_sentiment.columns if col.endswith('_sentiment_mean_cohort')]
        if not cohort_sentiment_cols:
            print(f"No weekly sentiment columns found in cohort sentiment data for year {target_year}.")
            plt.close(fig)
            return

        all_cohort_weeks = sorted(list(set([float(col.split('_')[0].replace('week', '')) for col in cohort_sentiment_cols])))

        cohort_ids = sorted(cohort_sentiment.index.unique())
        num_cohorts = len(cohort_ids)
        # FIX: Use matplotlib.colormaps.get_cmap directly
        cohort_colors = plt.colormaps.get_cmap('Dark2') # Corrected line
        cohort_linestyles = ['-', '--', '-.', ':', (0, (3, 5, 1, 5, 1, 5)), (0, (3, 1, 1, 1))]

        for i, cohort_id in enumerate(cohort_ids):
            row = cohort_sentiment.loc[cohort_id]
            sentiment_data = []
            weeks_for_this_cohort_plot = []
            # Ensure plot_type is defined for f-string, CohortSentimentPlotter always plots sentiment
            plot_type_local = 'sentiment'
            for wk in all_cohort_weeks:
                col_name = f"week{wk}_{plot_type_local}_mean_cohort"
                if col_name in row.index and pd.notna(row[col_name]):
                    sentiment_data.append(row[col_name])
                    weeks_for_this_cohort_plot.append(wk)
                else:
                    # Append NaN if data is missing for a week, so the line breaks
                    sentiment_data.append(np.nan)
                    weeks_for_this_cohort_plot.append(wk)

            # Filter out NaNs if plotting lines, but keep weeks_for_this_cohort_plot consistent for x-axis
            # For plotting, it's better to filter pairs of (week, data) rather than just data
            plot_pairs = [(w, s) for w, s in zip(weeks_for_this_cohort_plot, sentiment_data) if pd.notna(s)]
            if plot_pairs:
                plot_weeks, plot_sentiments = zip(*plot_pairs)
                plt.plot(plot_weeks, plot_sentiments,
                         marker='o', alpha=0.6,
                         linestyle=cohort_linestyles[i % len(cohort_linestyles)],
                         color=cohort_colors(i),
                         label=f'Cohort {cohort_id}')

        plt.title(f"Weekly Cohort Sentiment (Year {target_year})")
        plt.xlabel("Week")
        plt.ylabel("Average Sentiment (Mean)")
        plt.axhline(0, color='grey', linestyle='--')
        plt.ylim(-1, 1)
        plt.grid(True)
        plt.xticks([int(w) for w in all_cohort_weeks])
        plt.legend(title='Cohort ID', bbox_to_anchor=(1.0, 1), loc='upper left')
        plt.tight_layout(pad=0.5)
        pdf_pages.savefig(fig)
        plt.close(fig)
        print(f"\u2705 Weekly cohort sentiment plot for year {target_year} saved to PDF.")


class XCultureVisualizer:
    def __init__(self, plotting_strategies: list[IPlottingStrategy], team_summary, top_10_teams_summary, lowest_10_teams_summary, noun_phrases, yake_keywords, vader_tension_phrases, output_pdf_file, target_year=None, target_top_noun_phrases=30,
                 clustered_df_target_k=pd.DataFrame(), cluster_centroids_target_k=pd.DataFrame(),
                 clustered_df_optimal_k=pd.DataFrame(), cluster_centroids_optimal_k=pd.DataFrame(),
                 kmeans_week1_col=None, kmeans_week2_col=None,
                 wcss_scores={}, silhouette_scores={}, optimal_k_elbow=None, optimal_k_silhouette=None,
                 target_k_clusters=3, cohort_sentiment=pd.DataFrame(),
                 id_col: str = 'team_id', id_col_type: str = 'Team'): # Added id_col and id_col_type

        self.plotting_strategies = plotting_strategies
        self.team_summary = team_summary
        self.top_10_teams_summary = top_10_teams_summary
        self.lowest_10_teams_summary = lowest_10_teams_summary
        self.noun_phrases = noun_phrases
        self.yake_keywords = yake_keywords
        self.vader_tension_phrases = vader_tension_phrases
        self.cohort_sentiment = cohort_sentiment
        self.output_pdf_file = output_pdf_file
        self.target_year = target_year
        self.target_top_noun_phrases = target_top_noun_phrases

        # Generic clustering results (for the selected clustering_target_level)
        self.clustered_df_target_k = clustered_df_target_k
        self.cluster_centroids_target_k = cluster_centroids_target_k
        self.clustered_df_optimal_k = clustered_df_optimal_k
        self.cluster_centroids_optimal_k = cluster_centroids_optimal_k
        self.kmeans_week1_col = kmeans_week1_col
        self.kmeans_week2_col = kmeans_week2_col
        self.wcss_scores = wcss_scores
        self.silhouette_scores = silhouette_scores
        self.optimal_k_elbow = optimal_k_elbow
        self.optimal_k_silhouette = optimal_k_silhouette

        self.id_col = id_col # Store the ID column name (e.g., 'team_id', 'student_id')
        self.id_col_type = id_col_type # Store the type (e.g., 'Team', 'Individual')

        self.target_k_clusters = target_k_clusters
        self.weeks_to_plot = self._derive_weeks_to_plot()
        self.global_max_word_count = self._calculate_global_max_word_count()

        if not self.team_summary.empty and self.weeks_to_plot:
            # print("Initialized: Visualizer. Ready to plot.")
            pass
        else:
            print("Visualizer initialized but no data or weeks to plot.")

    def _derive_weeks_to_plot(self):
        if self.team_summary.empty:
            return []
        available_sentiment_mean_weeks = sorted(list(set(
            [float(col.split('_')[0].replace('week', '')) for col in self.team_summary.columns if col.endswith('_sentiment_mean')] )))
        available_subjectivity_mean_weeks = sorted(list(set(
            [float(col.split('_')[0].replace('week', '')) for col in self.team_summary.columns if col.endswith('_subjectivity_mean')] )))
        available_word_count_mean_weeks = sorted(list(set(
            [float(col.split('_')[0].replace('week', '')) for col in self.team_summary.columns if col.endswith('_word_count_mean')] )))
        return sorted(list(set(available_sentiment_mean_weeks + available_subjectivity_mean_weeks + available_word_count_mean_weeks)))

    def _calculate_global_max_word_count(self):
        combined_df = pd.concat([self.top_10_teams_summary, self.lowest_10_teams_summary]).drop_duplicates(subset=['team_id'])
        if combined_df.empty:
            return 1
        word_count_cols = [col for col in combined_df.columns if col.endswith('_word_count_mean')]
        if not word_count_cols:
            return 1
        max_wc = combined_df[word_count_cols].max().max()
        if pd.isna(max_wc):
            return 1
        return max_wc * 1.1

    def plot_all_sentiment_and_subjectivity(self):
        if not self.plotting_strategies:
            print("No plotting strategies provided to the Visualizer.")
            return
        # Removed the empty check for team_summary here, as individual plots might still run.

        with PdfPages(self.output_pdf_file) as pdf:
            for strategy in self.plotting_strategies:
                plot_data = {
                    'team_summary': self.team_summary,
                    'top_10_teams_summary': self.top_10_teams_summary,
                    'lowest_10_teams_summary': self.lowest_10_teams_summary,
                    'noun_phrases': self.noun_phrases,
                    'yake_keywords': self.yake_keywords,
                    'vader_tension_phrases': self.vader_tension_phrases,
                    'cohort_sentiment': self.cohort_sentiment,
                    'target_year': self.target_year,
                    'target_top_noun_phrases': self.target_top_noun_phrases,

                    # Generic clustering data
                    'clustered_df': self.clustered_df_target_k, # For target_k
                    'centroids_df': self.cluster_centroids_target_k, # For target_k
                    'clustered_df_optimal_k': self.clustered_df_optimal_k, # For optimal_k
                    'centroids_df_optimal_k': self.cluster_centroids_optimal_k, # For optimal_k
                    'kmeans_week1_col': self.kmeans_week1_col,
                    'kmeans_week2_col': self.kmeans_week2_col,
                    'wcss_scores': self.wcss_scores,
                    'silhouette_scores': self.silhouette_scores,
                    'optimal_k_elbow': self.optimal_k_elbow,
                    'optimal_k_silhouette': self.optimal_k_silhouette,
                    'id_col': self.id_col,
                    'id_col_type': self.id_col_type,

                    'target_k_clusters': self.target_k_clusters,
                    'weeks_to_plot': self.weeks_to_plot,
                    'global_max_word_count': self.global_max_word_count,
                    'dataframe': pd.DataFrame(),
                    'title': '',
                    'plot_type': ''
                }

                if isinstance(strategy, SentimentAndSubjectivityPlotter):
                    for df_type, df, suffix in [
                        ('top_10', self.top_10_teams_summary, "Highest Mean Sentiment Teams"),
                        ('lowest_10', self.lowest_10_teams_summary, "Lowest Mean Sentiment Teams")
                    ]:
                        if not df.empty:
                            plot_data.update({
                                'dataframe': df,
                                'title': f"Weekly Comments Sentiment - {suffix}",
                                'plot_type': 'sentiment'
                            })
                            strategy.plot(pdf, plot_data)
                            plot_data.update({
                                'dataframe': df,
                                'title': f"Weekly Comments Subjectivity - {suffix}",
                                'plot_type': 'subjectivity'
                            })
                            strategy.plot(pdf, plot_data)
                elif isinstance(strategy, SentimentWordCountPlotter):
                    for df_type, df, suffix in [
                        ('top_10', self.top_10_teams_summary, "Highest Mean Sentiment Teams"),
                        ('lowest_10', self.lowest_10_teams_summary, "Lowest Mean Sentiment Teams")
                    ]:
                        if not df.empty:
                            plot_data.update({
                                'dataframe': df,
                                'title': f"Weekly Comments Sentiment & Word Count - {suffix}",
                                'plot_type': 'sentiment' # Added plot_type here for CohortSentimentPlotter
                            })
                            strategy.plot(pdf, plot_data)
                elif isinstance(strategy, NounPhraseReporter):
                    strategy.plot(pdf, plot_data)
                elif isinstance(strategy, YAKEKeywordReporter):
                    strategy.plot(pdf, plot_data)
                elif isinstance(strategy, VADERCompoundPhraseReporter):
                    strategy.plot(pdf, plot_data)
                elif isinstance(strategy, KMeansClusterPlotter):
                    # Plot for target k
                    if not self.clustered_df_target_k.empty and not self.cluster_centroids_target_k.empty:
                        plot_data.update({
                            'clustered_df': self.clustered_df_target_k,
                            'centroids_df': self.cluster_centroids_target_k,
                            'title_suffix': f'({self.id_col_type} k={self.target_k_clusters})',
                            'kmeans_week1_col': self.kmeans_week1_col,
                            'kmeans_week2_col': self.kmeans_week2_col,
                            'id_col': self.id_col,
                            'id_col_type': self.id_col_type
                        })
                        strategy.plot(pdf, plot_data)
                    else:
                         print(f"Skipping {self.id_col_type} K-Means plot for target k ({self.target_k_clusters}) for year {self.target_year} due to no valid data.")

                    if self.optimal_k_silhouette is not None and not self.clustered_df_optimal_k.empty and not self.cluster_centroids_optimal_k.empty:
                        plot_data.update({
                            'clustered_df': self.clustered_df_optimal_k,
                            'centroids_df': self.cluster_centroids_optimal_k,
                            'title_suffix': f'({self.id_col_type} Optimal k={int(self.optimal_k_silhouette)} - Silhouette Method)',
                            'kmeans_week1_col': self.kmeans_week1_col,
                            'kmeans_week2_col': self.kmeans_week2_col,
                            'id_col': self.id_col,
                            'id_col_type': self.id_col_type
                        })
                        strategy.plot(pdf, plot_data)
                    else:
                        print(f"Skipping {self.id_col_type} K-Means plot for optimal k (Silhouette method) for year {self.target_year} due to no valid data or k.")

                elif isinstance(strategy, WCSSSilhouettePlotter):
                    # Plot for current clustering level
                    plot_data.update({
                        'wcss_scores': self.wcss_scores,
                        'silhouette_scores': self.silhouette_scores,
                        'optimal_k_elbow': self.optimal_k_elbow,
                        'optimal_k_silhouette': self.optimal_k_silhouette,
                        'id_col_type': self.id_col_type
                    })
                    strategy.plot(pdf, plot_data)

                elif isinstance(strategy, CohortSentimentPlotter):
                    # CohortSentimentPlotter only plots sentiment, so plot_type is implicitly 'sentiment'
                    plot_data.update({'plot_type': 'sentiment'}) # Explicitly set plot_type for this plotter
                    strategy.plot(pdf, plot_data)

        print(f"\u2705 All plots successfully saved to {self.output_pdf_file}")


class MainProcessor:
    def __init__(self, full_comments_df, codebook_df, var_week_map, weeks, text_analyzers: list[ITextAnalyzer],
                 team_clustering_strategy: IClusteringStrategy, individual_clustering_strategy: IClusteringStrategy,
                 team_summary_output, combined_top_lowest_output, output_pdf_file,
                 target_year=None, target_top_noun_phrases=30, target_k_clusters=3, k_range_for_optimizer=None,
                 plotting_strategies: list[IPlottingStrategy]=None, clustering_target_level: str = 'team'):

        self.codebook_df = codebook_df
        self.var_week_map = var_week_map
        self.weeks = weeks
        self.text_analyzers = text_analyzers
        self.team_clustering_strategy = team_clustering_strategy
        self.individual_clustering_strategy = individual_clustering_strategy
        self.team_summary_output = team_summary_output
        self.combined_top_lowest_output = combined_top_lowest_output
        self.output_pdf_file = output_pdf_file
        self.target_year = target_year
        self.target_k_clusters = target_k_clusters
        self.k_range_for_optimizer = k_range_for_optimizer if k_range_for_optimizer is not None else range(2,11)
        self.plotting_strategies = plotting_strategies if plotting_strategies is not None else []
        self.clustering_target_level = clustering_target_level # Store the new parameter

        # Initialize generic attributes for clustering results
        self.clustered_df_target_k = pd.DataFrame()
        self.cluster_centroids_target_k = pd.DataFrame()
        self.clustered_df_optimal_k = pd.DataFrame()
        self.cluster_centroids_optimal_k = pd.DataFrame()
        self.wcss_scores = {}
        self.silhouette_scores = {}
        self.optimal_k_elbow = None
        self.optimal_k_silhouette = None
        self.kmeans_week1_col = None
        self.kmeans_week2_col = None
        self.id_col = None
        self.id_col_type = None

        self.data_processor = XCultureData(
            full_comments_df,
            self.codebook_df,
            self.var_week_map,
            self.weeks,
            self.text_analyzers,
            self.target_year,
            target_top_noun_phrases
        )

        self.team_analyzer = XCultureTeamAnalysis(
            self.data_processor.comments_df,
            self.data_processor.var_week_map,
            self.data_processor.weeks
        )

        week1_sentiment_col = 'week1.0_sentiment_mean'
        week10_col_option = 'week10.0_sentiment_mean'
        week11_col_option = 'week11.0_sentiment_mean'
        selected_weekX_sentiment_col_hint = None # This will be either week10 or week11 mean col name

        team_summary_df = self.team_analyzer.team_summary

        # Get the nationality results for the current year after data_processor has run its analysis
        current_year_nationality_results = self.data_processor.yearly_nationality_results.get(self.target_year, pd.DataFrame())

        print(current_year_nationality_results.columns.tolist())
        print(f'nationality{self.target_year}')
        team_current_year_nationality_list = current_year_nationality_results.groupby('team_id')["nationality"].agg(list).reset_index()
        team_summary_df = pd.merge(team_summary_df, team_current_year_nationality_list, on='team_id', how='outer')

        # Determine selected_weekX_sentiment_col_hint based on team_summary_df data availability
        has_week10 = week10_col_option in team_summary_df.columns
        has_week11 = week11_col_option in team_summary_df.columns

        count_week10 = team_summary_df[week10_col_option].count() if has_week10 else -1
        count_week11 = team_summary_df[week11_col_option].count() if has_week11 else -1

        if team_summary_df.empty or (not has_week10 and not has_week11):
            print("Warning: team_summary_df is empty or neither week10 nor week11 sentiment columns exist. Using week1.0_sentiment_mean as fallback for weekX selection hint.")
            selected_weekX_sentiment_col_hint = week1_sentiment_col # Fallback
        elif count_week10 >= count_week11:
            selected_weekX_sentiment_col_hint = week10_col_option
            print(f"Selected '{week10_col_option}' for K-Means clustering (non-nulls: {count_week10}).")
        else:
            selected_weekX_sentiment_col_hint = week11_col_option
            print(f"Selected '{week11_col_option}' for K-Means clustering (non-nulls: {count_week11}).")

        # Ensure week1_sentiment_col exists, if not, use the selected_weekX_sentiment_col_hint as fallback for consistency in clustering.
        actual_week1_col = week1_sentiment_col
        if week1_sentiment_col not in team_summary_df.columns:
            print(f"Warning: '{week1_sentiment_col}' not found for K-Means clustering. Using '{selected_weekX_sentiment_col_hint}' for both axes as a fallback for initial week selection.")
            actual_week1_col = selected_weekX_sentiment_col_hint

        clustering_results = {}
        if self.clustering_target_level == 'team':
            self.id_col = 'team_id'
            self.id_col_type = 'Team'
            self.kmeans_week1_col = actual_week1_col
            self.kmeans_week2_col = selected_weekX_sentiment_col_hint

            clustering_results = self.team_clustering_strategy.perform_clustering(
                data_df=self.team_analyzer.team_summary,
                week1_sentiment_col=self.kmeans_week1_col,
                weekX_sentiment_col=self.kmeans_week2_col,
                target_k_clusters=self.target_k_clusters,
                k_range_optimizer=self.k_range_for_optimizer,
                var_week_map=self.var_week_map
            )
        elif self.clustering_target_level == 'individual':
            self.id_col = 'student_id'
            self.id_col_type = 'Individual'
            clustering_results = self.individual_clustering_strategy.perform_clustering(
                data_df=self.data_processor.comments_df, # Use comments_df for individual clustering
                week1_sentiment_col=actual_week1_col, # Hint for week 1
                weekX_sentiment_col=selected_weekX_sentiment_col_hint, # Hint for week X
                target_k_clusters=self.target_k_clusters,
                k_range_optimizer=self.k_range_for_optimizer,
                var_week_map=self.var_week_map,
                cohort_sentiment_data=self.data_processor.yearly_cohort_sentiment.get(self.target_year, pd.DataFrame())
            )
            self.kmeans_week1_col = clustering_results.get('kmeans_week1_col')
            self.kmeans_week2_col = clustering_results.get('kmeans_week2_col')
        else:
            print(f"Warning: Unknown clustering_target_level '{self.clustering_target_level}'. No clustering performed.")

        # Assign clustering results to generic attributes
        self.clustered_df_target_k = clustering_results.get('clustered_df_target_k', pd.DataFrame())
        self.cluster_centroids_target_k = clustering_results.get('cluster_centroids_target_k', pd.DataFrame())
        self.clustered_df_optimal_k = clustering_results.get('clustered_df_optimal_k', pd.DataFrame())
        self.cluster_centroids_optimal_k = clustering_results.get('cluster_centroids_optimal_k', pd.DataFrame())
        self.wcss_scores = clustering_results.get('wcss_scores', {})
        self.silhouette_scores = clustering_results.get('silhouette_scores', {})
        self.optimal_k_elbow = clustering_results.get('optimal_k_elbow', None)
        self.optimal_k_silhouette = clustering_results.get('optimal_k_silhouette', None)

        team_analysis_end_time = datetime.now()
        print(f"MainProcessor X-Culture Analysis Finished: {team_analysis_end_time}")

        self.exporter = XCultureExporter(self.target_year)

    @staticmethod
    def _validate_target_years(comments_df, initial_years_list):
        if 'analysis_year' in comments_df.columns:
            available_years = comments_df['analysis_year'].dropna().astype(int).unique()
            validated_years = [year for year in initial_years_list if year in available_years]
            if len(validated_years) < len(initial_years_list):
                skipped_years = set(initial_years_list) - set(validated_years)
                print(f"Warning: Some target years ({skipped_years}) were not found in the data and will be skipped.")
            if not validated_years:
                print("Error: No valid target years found in the data after validation. Exiting process.")
            return validated_years
        else:
            print("Warning: 'analysis_year' column not available for validating target_years_list. All specified years will be attempted.")
            return initial_years_list

    def run_export_and_plot(self):
        self.exporter.export_dataframe_to_excel(self.team_analyzer.team_summary, self.team_summary_output, "Team_Summary")

        engagement_summ_additional_dataframes = {}

        # Add nationality data for export based on the selected level
        nationality_data = self.data_processor.yearly_nationality_results.get(self.target_year, pd.DataFrame())
        if not nationality_data.empty:
            nationality_df = nationality_data.copy()
            nationality_df.columns = ['student_id', 'team_id', 'Nationality']
            # engagement_summ_additional_dataframes["Yearly_Nationality_Result"] = nationality_df
            # print(f"DEBUG: Sample nationality_df entries: {list(nationality_df.items())[:10]}")
        else:
            print(f"Warning: No nationality results found for year {self.target_year}.")

        if not self.data_processor.comments_df.empty:
            ordered_comment_related_cols = []

            vars_by_week = defaultdict(list)
            for var_name, week_num in self.data_processor.var_week_map.items():
                if var_name in self.data_processor.comments_df.columns:
                    vars_by_week[week_num].append(var_name)

            for week in sorted(vars_by_week.keys()):
                for comment_var in sorted(vars_by_week[week]):
                    if f"{comment_var}_word_count" in self.data_processor.comments_df.columns:
                        ordered_comment_related_cols.append(f"{comment_var}_word_count")
                    if f"{comment_var}_sentiment" in self.data_processor.comments_df.columns:
                        ordered_comment_related_cols.append(f"{comment_var}_sentiment")
                    if f"{comment_var}_subjectivity" in self.data_processor.comments_df.columns:
                        ordered_comment_related_cols.append(f"{comment_var}_subjectivity")
                    if f"{comment_var}_vader" in self.data_processor.comments_df.columns:
                        ordered_comment_related_cols.append(f"{comment_var}_vader")
                    if f"{comment_var}_vader_neg" in self.data_processor.comments_df.columns:
                        ordered_comment_related_cols.append(f"{comment_var}_vader_neg")
                    if f"{comment_var}_vader_neu" in self.data_processor.comments_df.columns:
                        ordered_comment_related_cols.append(f"{comment_var}_vader_neu")
                    if f"{comment_var}_vader_pos" in self.data_processor.comments_df.columns:
                        ordered_comment_related_cols.append(f"{comment_var}_vader_pos")


                if f"week{week}_word_count" in self.data_processor.comments_df.columns:
                    ordered_comment_related_cols.append(f"week{week}_word_count")
                if f"week{week}_sentiment" in self.data_processor.comments_df.columns:
                    ordered_comment_related_cols.append(f"week{week}_sentiment")
                if f"week{week}_subjectivity" in self.data_processor.comments_df.columns:
                    ordered_comment_related_cols.append(f"week{week}_subjectivity")
                if f"week{week}_vader" in self.data_processor.comments_df.columns:
                    ordered_comment_related_cols.append(f"week{week}_vader")
                if f"week{week}_vader_neg" in self.data_processor.comments_df.columns:
                    ordered_comment_related_cols.append(f"week{week}_vader_neg")
                if f"week{week}_vader_neu" in self.data_processor.comments_df.columns:
                    ordered_comment_related_cols.append(f"week{week}_vader_neu")
                if f"week{week}_vader_pos" in self.data_processor.comments_df.columns:
                    ordered_comment_related_cols.append(f"week{week}_vader_pos")

            fixed_cols = [
                        'student_id', 'team_id', 'cohort_id', 'personal_desc',
                        'comment_at1', 'comment_at2', 'comment_at3', 'comment_at4', 'comment_at5',
                        'comment_at6', 'comment_at7', 'comment_at8', 'comment_at9', 'comment_at10',
                        'comment_at11', 'changes', 'typos_at1'
                        ]
            for col in fixed_cols:
                if col in self.data_processor.comments_df.columns:
                    ordered_comment_related_cols.append(col)

            individual_comments_df = self.data_processor.comments_df[ordered_comment_related_cols].copy()

            # Create single merged df for Comments Analysis and Nationality.
            merged_analysis_nationality_df = pd.merge(nationality_df, individual_comments_df, on='student_id', how='outer')
            engagement_summ_additional_dataframes["Nat_and_Analysis_and_Text"] = merged_analysis_nationality_df

        vader_tension_data = self.data_processor.yearly_vader_tension_phrases.get(self.target_year, [])
        if vader_tension_data:
            vader_df = pd.DataFrame(vader_tension_data, columns=['Phrase', 'VADER Compound Score', 'Frequency'])
            engagement_summ_additional_dataframes["VADER_Tension_Phrases"] = vader_df

        # Add clustering results for export based on the selected level
        if self.clustering_target_level == 'team':
            if self.wcss_scores:
                wcss_df = pd.DataFrame(self.wcss_scores.items(), columns=['k', 'WCSS'])
                engagement_summ_additional_dataframes["Team_WCSS_Scores"] = wcss_df
            if self.silhouette_scores:
                silhouette_df = pd.DataFrame(self.silhouette_scores.items(), columns=['k', 'Silhouette_Score'])
                engagement_summ_additional_dataframes["Team_Silhouette_Scores"] = silhouette_df
            if not self.clustered_df_target_k.empty:
                engagement_summ_additional_dataframes["Team_Clusters_Target_K"] = self.clustered_df_target_k
            if not self.cluster_centroids_target_k.empty:
                engagement_summ_additional_dataframes["Team_Centroids_Target_K"] = self.cluster_centroids_target_k
            if not self.clustered_df_optimal_k.empty:
                engagement_summ_additional_dataframes["Team_Clusters_Opt_K"] = self.clustered_df_optimal_k
            if not self.cluster_centroids_optimal_k.empty:
                engagement_summ_additional_dataframes["Team_Centroids_Opt_K"] = self.cluster_centroids_optimal_k

        elif self.clustering_target_level == 'individual':
            if self.wcss_scores:
                wcss_df = pd.DataFrame(self.wcss_scores.items(), columns=['k', 'WCSS'])
                engagement_summ_additional_dataframes["Indiv_WCSS_Scores"] = wcss_df
            if self.silhouette_scores:
                silhouette_df = pd.DataFrame(self.silhouette_scores.items(), columns=['k', 'Silhouette_Score'])
                engagement_summ_additional_dataframes["Indiv_Silhouette_Scores"] = silhouette_df
            if not self.clustered_df_target_k.empty:
                engagement_summ_additional_dataframes["Indiv_Clusters_Target_K"] = self.clustered_df_target_k
            if not self.cluster_centroids_target_k.empty:
                engagement_summ_additional_dataframes["Indiv_Centroids_Target_K"] = self.cluster_centroids_target_k
            if not self.clustered_df_optimal_k.empty:
                engagement_summ_additional_dataframes["Indiv_Clusters_Opt_K"] = self.clustered_df_optimal_k
            if not self.cluster_centroids_optimal_k.empty:
                engagement_summ_additional_dataframes["Indiv_Centroids_Opt_K"] = self.cluster_centroids_optimal_k

        cohort_sentiment_df = self.data_processor.yearly_cohort_sentiment.get(self.target_year, pd.DataFrame())
        if not cohort_sentiment_df.empty:
            engagement_summ_additional_dataframes["Cohort_Sentiment"] = cohort_sentiment_df

        # ---------------------------------
        # *** Main Export to Excel Step ***
        # ---------------------------------
        if engagement_summ_additional_dataframes:
            self.exporter.export_multiple_dataframes_to_excel(engagement_summ_additional_dataframes, self.team_summary_output)

        combined_dataframes_for_export = {
            "Top_10_Sentiment": self.team_analyzer.top_10_teams_summary,
            "Low_10_Sentiment": self.team_analyzer.lowest_10_teams_summary
        }

        self.exporter.export_multiple_dataframes_to_excel(combined_dataframes_for_export, self.combined_top_lowest_output)

        visualizer_instance = XCultureVisualizer(
            self.plotting_strategies,
            self.team_analyzer.team_summary,
            self.team_analyzer.top_10_teams_summary,
            self.team_analyzer.lowest_10_teams_summary,
            self.data_processor.yearly_top_noun_phrases,
            self.data_processor.yearly_yake_keywords,
            self.data_processor.yearly_vader_tension_phrases,
            self.output_pdf_file,
            self.target_year,
            target_top_noun_phrases,
            clustered_df_target_k=self.clustered_df_target_k,
            cluster_centroids_target_k=self.cluster_centroids_target_k,
            clustered_df_optimal_k=self.clustered_df_optimal_k,
            cluster_centroids_optimal_k=self.cluster_centroids_optimal_k,
            kmeans_week1_col=self.kmeans_week1_col,
            kmeans_week2_col=self.kmeans_week2_col,
            wcss_scores=self.wcss_scores,
            silhouette_scores=self.silhouette_scores,
            optimal_k_elbow=self.optimal_k_elbow,
            optimal_k_silhouette=self.optimal_k_silhouette,
            target_k_clusters=self.target_k_clusters,
            cohort_sentiment=self.data_processor.yearly_cohort_sentiment.get(self.target_year, pd.DataFrame()),
            id_col=self.id_col,
            id_col_type=self.id_col_type
        )
        visualizer_instance.plot_all_sentiment_and_subjectivity()


class XCultureYearProcessor:
    def __init__(self, full_comments_df, global_codebook_df, global_var_week_map, global_weeks, text_analyzers: list[ITextAnalyzer],
                 team_clustering_strategy: IClusteringStrategy, individual_clustering_strategy: IClusteringStrategy,
                 plotting_strategies: list[IPlottingStrategy], output_data_dir, output_plots_dir, target_year,
                 target_top_noun_phrases=30, target_k_clusters=3, k_range_for_optimizer=None,
                 clustering_target_level: str = 'team'):
        self.full_comments_df = full_comments_df
        self.global_codebook_df = global_codebook_df
        self.global_var_week_map = global_var_week_map
        self.global_weeks = global_weeks
        self.text_analyzers = text_analyzers
        self.team_clustering_strategy = team_clustering_strategy
        self.individual_clustering_strategy = individual_clustering_strategy
        self.plotting_strategies = plotting_strategies
        self.output_data_dir = output_data_dir
        self.output_plots_dir = output_plots_dir
        self.target_year = target_year
        self.target_top_noun_phrases = target_top_noun_phrases
        self.target_k_clusters = target_k_clusters
        self.k_range_for_optimizer = k_range_for_optimizer
        self.clustering_target_level = clustering_target_level # Store the new parameter

    def process_year(self):
        print(f"\n--- Processing year: {self.target_year} ---")
        timenow_str = datetime.now().strftime('%Y%m%d%H%M')

        all_teams_summary_output_file = f"{self.output_data_dir}{self.target_year}_X-Culture_Engagement_Summ_{timenow_str}.xlsx"
        sample_teams_summary_output_file = f"{self.output_data_dir}{self.target_year}_X-Culture_Sample_Teams_Summ_{timenow_str}.xlsx"
        plots_pdf_output_file = f"{self.output_plots_dir}{self.target_year}_X-Culture_Visulization_{timenow_str}.pdf"

        processor = MainProcessor(
            self.full_comments_df,
            self.global_codebook_df,
            self.global_var_week_map,
            self.global_weeks,
            self.text_analyzers,
            self.team_clustering_strategy,
            self.individual_clustering_strategy,
            all_teams_summary_output_file,
            sample_teams_summary_output_file,
            plots_pdf_output_file,
            self.target_year,
            self.target_top_noun_phrases,
            self.target_k_clusters,
            self.k_range_for_optimizer,
            self.plotting_strategies,
            self.clustering_target_level # Pass the clustering_target_level
        )
        processor.run_export_and_plot()



class OSDriveCheck:
    def check_and_mount_drive(self):
        if not os.path.exists('/content/drive'):
            print('Google Drive is not mounted. Mounting now...')
            drive.mount('/content/drive/')
        else:
            print('Google Drive is already mounted.')


# -------------------------------------------------------------------
# Container Process for X-Culture Data Analysis
# -------------------------------------------------------------------

# target_years_list = [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
# target_years_list = [2020, 2021, 2022]
target_years_list = [2019]
target_top_noun_phrases = 100
target_k_clusters = 4
k_range_for_optimizer = range(2, 11)

# New global variable to select clustering level
clustering_level_selection = 'individual' # Can be 'team' or 'individual'

input_data_dir = "/content/drive/MyDrive/x-culture/01_Data/"
output_data_dir = "/content/drive/MyDrive/x-culture/02_Results_Data/"
output_plots_dir = "/content/drive/MyDrive/x-culture/02_Results_Plots/"

codebook_file = f"{input_data_dir}Comments_CodeBook_v4.xlsx"
comments_file = f"{input_data_dir}Comments_Data_Original.xlsx"
geo_json = f"{input_data_dir}dict_common_countries_full.json"
print(f"Defined geo_json path: {geo_json}")

drive_checker = OSDriveCheck()
drive_checker.check_and_mount_drive()

main_start_time = datetime.now()
print(f"Container process started at: {main_start_time}")

codebook_processor = XCultureCodebookProcessor(codebook_file)
global_codebook_df = codebook_processor.codebook_df
global_var_week_map = codebook_processor.var_week_map
global_weeks = codebook_processor.weeks

full_comments_df = pd.read_excel(comments_file, engine="openpyxl")
print(f"Comments file '{comments_file}' loaded. Total rows: {len(full_comments_df)}")
if 'student_id' in full_comments_df.columns:
    full_comments_df['analysis_year'] = pd.to_numeric(
        full_comments_df['student_id'].astype(str).str.split('-').str[0],
        errors='coerce'
    )
    print("Pre-extracted 'analysis_year' from 'student_id' column.")
else:
    print("Warning: 'student_year' column not found in comments_df. Optimization skipped.")

target_years_list = MainProcessor._validate_target_years(full_comments_df, target_years_list)

# Instantiate concrete ITextAnalyzer implementations
text_analyzers = [
    XCultureNounPhraseExtractor(),
    XCultureYAKEKeywordExtractor(),
    XCultureVADERAnalyzer(),
    XCultureNationalityMiner(geo_json_file=geo_json) # Pass the geo_json_file
]

# Instantiate concrete IClusteringStrategy implementations
team_clustering_strategy = TeamSentimentKMeansClusteringStrategy()
individual_clustering_strategy = IndividualSentimentKMeansClusteringStrategy()

# Instantiate concrete IPlottingStrategy implementations
plotting_strategies = [
    SentimentAndSubjectivityPlotter(),
    SentimentWordCountPlotter(),
    NounPhraseReporter(),
    YAKEKeywordReporter(),
    VADERCompoundPhraseReporter(),
    KMeansClusterPlotter(),
    WCSSSilhouettePlotter(),
    CohortSentimentPlotter()
]

for year in target_years_list:
    year_processor = XCultureYearProcessor(
        full_comments_df,
        global_codebook_df,
        global_var_week_map,
        global_weeks,
        text_analyzers,
        team_clustering_strategy,
        individual_clustering_strategy,
        plotting_strategies,
        output_data_dir,
        output_plots_dir,
        year,
        target_top_noun_phrases,
        target_k_clusters,
        k_range_for_optimizer,
        clustering_target_level=clustering_level_selection # Pass the selected clustering level
    )
    year_processor.process_year()

main_end_time = datetime.now()
duration = main_end_time - main_start_time
print(f"Container process finished at {main_end_time}, Duration {duration}")
```
